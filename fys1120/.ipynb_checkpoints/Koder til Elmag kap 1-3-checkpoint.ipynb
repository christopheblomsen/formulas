# Elektrisk felt uke 1

## Felles for alt

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Funksjon for elektrisk felt
$$ \mathbf{E} = \frac{\mathbf{F}}{q}= \frac{Q \mathbf{R}}{4\pi \varepsilon_0 R^2} $$

```python
def efield(r,qi,ri):
    # 1/(4*pi*epsilon_0) = 1
    Ri = r - ri
    Rinorm = np.linalg.norm(Ri)
    return qi*Ri/Rinorm**3
```

#### Med eksempel

```python
L = 2
N = 30
x = np.linspace(-L,L,N)
y = np.linspace(-L,L,N)
rx,ry = np.meshgrid(x,y)
q1 = -1.0
r1 = np.array([0,0])
q2 = 1.0
r2 = np.array([1,0])
Ex = np.zeros((N,N),float)
Ey = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],ry.flat[i]])
    Ex.flat[i],Ey.flat[i] = efield(r,q1,r1) + efield(r,q2,r2)
plt.figure(figsize=(10,10))
plt.quiver(rx,ry,Ex,Ey)
plt.axis('equal')
```

# Elektrisk potensial uke 2

## Elektrisk dipol
$$ V(\vec{r}) = \sum\limits_i \frac{Q_i}{4\pi\varepsilon_0 R_i} $$

```python
def epotlist(r,Q,R):
    V = 0
    for i in range(len(R)):
        Ri = r - R[i]
        Qi = Q[i]
        Rinorm = np.linalg.norm(Ri)
        V = V + Qi/Rinorm
    return V
```

```python
# Sette opp ladninger i liste
R = []
Q = []
r1 = np.array([1,0])
Q1 = 1
R.append(r1)
Q.append(Q1)
r2 = np.array([-1,0])
Q2 = -1
R.append(r2)
Q.append(Q2)

Lx = 3
Ly = 3
N = 21
x = np.linspace(-Lx,Lx,N)
y = np.linspace(-Ly,Ly,N)
rx,ry = np.meshgrid(x,y)

# Finne verdier for elektrisk potensiale på grid
V = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],ry.flat[i]])
    V.flat[i] = epotlist(r,Q,R)
plt.figure(figsize=(8,8))
plt.imshow(V)
plt.colorbar()
```

```python
plt.figure(figsize=(8,8))
plt.contourf(rx,ry,V,20)
plt.colorbar()
```

## Elektrisk felt
$$ \mathbf{E} = - \nabla V $$

```python
Ey,Ex = np.gradient(V)
Ey = -Ey
Ex = -Ex
plt.figure(figsize=(8,8))
plt.contourf(rx,ry,V,20)
plt.quiver(rx,ry,Ex,Ey)
```

## Ringladning

```python
def epotlist(r,Q,R):
    V = 0
    for i in range(len(R)):
        Ri = r - R[i]
        Qi = Q[i]
        Rinorm = np.linalg.norm(Ri)
        V = V + Qi/Rinorm
    return V
```

```python
# Lage ring
Q = []
R = []
M = 100
a = 2.0
q = 1.0
deltatheta = 2*np.pi/M
for i in range(M):
    thetai = i*deltatheta
    xi = a*np.cos(thetai)
    yi = a*np.sin(thetai)
    ri = np.array([xi,yi,0])
    R.append(ri)
    Qi = q/M
    Q.append(Qi)
Lx = 5
Lz = 5
N = 30
x = np.linspace(-Lx,Lx,N)
z = np.linspace(-Lz,Lz,N)
rx,rz = np.meshgrid(x,z)


V = np.zeros((N,N),float)
for i in range(len(rx.flat)):
    r = np.array([rx.flat[i],0,rz.flat[i]])
    V.flat[i] = epotlist(r,Q,R)

plt.figure(figsize=(8,8))
plt.contourf(V)
```

```python
Ez,Ex = np.gradient(-V)
plt.figure(figsize=(8,8))
plt.contourf(rx,rz,V)
plt.quiver(rx,rz,Ex,Ez)
```
