In electrostatics we started from Coulomb's law, which we used to find the electric field, and used this to derive Gauss' law, $\nabla \cdot D = \rho$, which is useful for symmetric situations (and for forming Maxwell's equations which we can solve numerically). In addition we found that the electric field (in electrostatics) is curl-free: $\nabla \times \mathbf{E} = 0$. These laws for the foundation of Maxwell's equations and have proved important tools for the development of magnetostatics.

In magnetostatics, we have so far introduced Biot-Savart's law to find the magnetic field. In this chapter, we will introduce Ampere's law, $\nabla \times \mathbf{B} = \mu_0 \mathbf{J}$, which is what corresponds to Gauss' law for magnetic fields. In addition, we will argue that the magnetic field is divergence free, $\nabla \cdot \mathbf{B} = 0$, which is equivalent to stating that there are no magnetic monopoles. In combination, these two laws for a powerful theoretical tool that allows us to find expressions for the magnetic field in situations with a high degree of symmetry. 

# Magnetic flux

For electric fields, we found Gauss' law, which related the flux of the electric field to the net free charge: $\int_S \mathbf{E} \cdot \mathrm{d} \mathbf{S} = Q_{\text{in}}/\epsilon_0$ on integral form and $\nabla \cdot \mathbf{E} = \rho/\epsilon_0$ on differential form. We could derive this law from Coulomb's law. Electric fields are generated by charges, whereas magnetic fields are generated by current elements. Do we still have a similar law for magnetic fields?

We define the magnetic flux through a surface $S$ as

<!-- Equation labels as ordinary links -->
<div id="_auto1"></div>

$$
\begin{equation}
\Phi_S = \int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} \; .
\label{_auto1} \tag{1}
\end{equation}
$$

For any *closed* surface, this flux is always zero! (We demonstrate this below). We can rewrite this on differential form using the divergence theorem, where we relate the flux of $\mathbf{B}$ through a closed surface $S$ to the volume integral of the divergence of $\mathbf{B}$ throught the volume $v$ enclosed by $S$:

<!-- Equation labels as ordinary links -->
<div id="_auto2"></div>

$$
\begin{equation}
\int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} = \int_v \nabla \cdot \mathbf{B} \mathrm{d} v = \; .
\label{_auto2} \tag{2}
\end{equation}
$$

Because this is true for any closed surface $S$, the argument of the integral must be equal to zero, $\nabla \cdot \mathbf{B} = 0$.

**Magnetic flux and the divergence of the magnetic field.**

The flux of the magnetic field $\mathbf{B}$ through any *closed* surface $S$ is always zero:

<!-- Equation labels as ordinary links -->
<div id="_auto3"></div>

$$
\begin{equation}
\oint_S  \mathbf{B} \cdot \mathrm{d} \mathbf{S} = 0 \; .
\label{_auto3} \tag{3}
\end{equation}
$$

This also means that the divergence of the magnetic field is zero:

<!-- Equation labels as ordinary links -->
<div id="_auto4"></div>

$$
\begin{equation}
\nabla \cdot \mathbf{B} = 0 \; .
\label{_auto4} \tag{4}
\end{equation}
$$

We do not know of any situations where these equations are not satisfied, that is, 'magnetic charge' or magnetic monopoles do not seem to exist.



## Proof that the divergence of the magnetic field is zero

Our plan is to demonstrate that the divergence of the magnetic field contribution from any current element is zero. Therefore, the sum of many such current elements will also have zero divergence. We look at a current element $I \mathrm{d} \mathbf{l}$. The contribution to the magnetic field $\mathrm{d} \mathbf{B}(\mathbf{r})$ in the point $\mathbf{r}$ from this current element is given by Biot-Savart-s law:

<!-- Equation labels as ordinary links -->
<div id="_auto5"></div>

$$
\begin{equation}
\mathrm{d} \mathbf{B} = \frac{\mu_0}{4 \pi} \frac{\mathbf{J}\mathrm{d} v \times \mathbf{\hat{R}}}{R^2} \; ,
\label{_auto5} \tag{5}
\end{equation}
$$

where $\mathbf{R} = \mathbf{r} - \mathbf{r}'$, where $\mathbf{r}'$ is the position of the current element. 
For simplicity we place the element in the origin so that $\mathbf{r}' = 0$. This means that the vector $\mathbf{R} = \mathbf{r} - \mathbf{r}' = \mathbf{r}$. The contribution $\mathrm{d} \mathbf{B}(\mathbf{r})$ from this current element to the magnetic field in $\mathbf{r}$ is then

<!-- Equation labels as ordinary links -->
<div id="_auto6"></div>

$$
\begin{equation}
\mathrm{d} \mathbf{B} = \frac{\mu_0}{4 \pi} \frac{\mathbf{J} \mathrm{d} v \times \mathbf{\hat{r}}}{r^2} \; ,
\label{_auto6} \tag{6}
\end{equation}
$$

We want to find the divergence of $\mathrm{d} \mathbf{B}$ and apply the product rule:

<!-- Equation labels as ordinary links -->
<div id="_auto7"></div>

$$
\begin{equation}
\nabla \cdot \left( \mathbf{J} \times \frac{\mathbf{\hat{r}}}{r^2} \right) = \frac{\mathbf{\hat{r}}}{r^2} \cdot \left( \nabla \times \mathbf{J} \right) - \mathbf{J} \cdot \left( \nabla \times \frac{\mathbf{\hat{r}}}{r^2} \right) \; .
\label{_auto7} \tag{7}
\end{equation}
$$

Notice that $\mathbf{J}$ here is the value for $\mathbf{J}$ in the origin. This value does not vary if we vary the reference point $\mathbf{r}$ and $\mathbf{J}$ is therefore a constant in the derivations. Therefore, $\nabla \times \mathbf{J}$ in this particular expression is zero. In addition, we see from the expression for the curl of a field in spherical coordinates that $\nabla \times \left( \mathbf{\hat{r}}/r^2 \right)$ always is zero. 
 We therefore conclude that the divergence of $\mathrm{d} \mathbf{B}$ is zero, and, using the superposition principle, that the divergence of $\mathbf{B}$ is zero. This implies that the magnetic flux through any closed surface is zero.

## Implications for a vector potential

We have previously found that $\nabla \times \mathbf{E} = 0$ allowed us to introduce a scalar potential $V$ so that $\mathbf{E} = - \nabla V$. Similarly, the observation that $\nabla \cdot \mathbf{B} = 0$ for magnetic field allows us to introduce a vector potential $\mathbf{A}$ so that $\mathbf{B} = \nabla \times \mathbf{A}$. Such a definition will automatically ensure that the divergence of $\mathbf{B}$ is zero, because the divergence of a curl always is zero.

How can we prove that it is possible to represent $\mathbf{B}$ as $\mathbf{B} = \nabla \times \mathbf{B}$? First, we demonstrate that there exists such a field $\mathbf{A}$ so that $\mathbf{B} = \nabla \times \mathbf{A}$. Second, we demonstrate that this field is unique up to a constant, just as for the scalar potential.

First, we demonstrate existence by introducing a possible solution [^ampereslaw-010]. We introduce the contribution from the vector potential from a current element $\mathbf{J}\mathrm{d} v$ in the origin (as we did above):

<!-- Equation labels as ordinary links -->
<div id="_auto8"></div>

$$
\begin{equation}
\mathrm{d} \mathbf{A} = \frac{\mu_0}{4 \pi} \frac{\mathbf{J}\mathrm{d} v}{r} \; .
\label{_auto8} \tag{8}
\end{equation}
$$

We show that this is consistent with Biot-Savart's law by finding $\nabla \times \mathrm{d} \mathbf{A}$. Here, we need to apply a product rule for curls (which is simply a rule for vector multiplication): $\nabla \times \left( a \mathbf{b} \right) = \nabla a \times \mathbf{b} + a \nabla \times \mathbf{b}$. In addition, we recall that $\nabla (1/r) = -\mathbf{\hat{r}}/r^2$ and that $\mathbf{J}$ is a constant in this derivative so that $\nabla \times \mathbf{J}\mathrm{d} v = 0$. We therefore get

<!-- Equation labels as ordinary links -->
<div id="_auto9"></div>

$$
\begin{equation}
\nabla \times \mathrm{d} \mathbf{A} = \frac{\mu_0}{4 \pi} \nabla \frac{1}{r}\times \left( \mathbf{J}\mathrm{d} v\right) = \frac{\mu_0}{4 \pi} \frac{- \mathbf{\hat{r}}}{r^2}\times \left( \mathbf{J}\mathrm{d} v \right) = \frac{\mu_0}{4 \pi}\frac{\mathbf{J}\mathrm{d} v \times \mathbf{\hat{r}}}{r^2} \; .
\label{_auto9} \tag{9}
\end{equation}
$$

We can then find the vector potential for a whole volume $v$ or for a circuit $C$ by applying the superposition principle, remembering that $\mathbf{R} = \mathbf{r} - \mathbf{r}'$, $\mathbf{r}'$ is the position of the current element, and that the integration is over $\mathbf{r}'$ (and not over $\mathbf{r}$!):

<!-- Equation labels as ordinary links -->
<div id="_auto10"></div>

$$
\begin{equation}
\mathbf{A} = \frac{\mu_0}{4 \pi} \int_v \frac{\mathbf{J}\mathrm{d} v}{R}
\label{_auto10} \tag{10}
\end{equation}
$$

or

<!-- Equation labels as ordinary links -->
<div id="_auto11"></div>

$$
\begin{equation}
\mathbf{A} = \frac{\mu_0}{4 \pi} \oint_C \frac{I\mathrm{d} \mathbf{l}}{R} \; .
\label{_auto11} \tag{11}
\end{equation}
$$

[^ampereslaw-010]: This requires us to know the solution. This is a cheating a bit, and does not really help you understand how we found this solution. Here, this is only used as a proof, but you may learn more about how to find vector potentials in other texts.

We will here not pursue the use of the vector potential, but only use it for a few theoretical derivations.


# Ampere's law

We found that electrostatics could be defined from the divergence and curl of the electric field, similarly, the laws of magnetostatics can be defined through the divergence and curl of the magnetic field. For the electric field, we have learned to apply Gauss' law to find the electric field in symmetric cases: $\nabla \cdot \mathbf{E} = \rho/\epsilon_0$. The corresponding law for magnetic fields is Ampere's law, which describes the curl of the magnetic field. We must therefore replace integration over closed surfaces for Gauss' law with integration over closed loops for Ampere's law.

Ampere's law on differential form states that the curl of the magnetic field is $\nabla \times \mathbf{B} = \mu_0 \mathbf{J}$. (We provide a proof of Ampere's law based on Biot-Savart's law below). We can transform the law to integral form using Stoke's theorem:

<!-- Equation labels as ordinary links -->
<div id="_auto12"></div>

$$
\begin{equation}
\int_S \left( \nabla \times \mathbf{B} \right) \cdot \mathrm{d} \mathbf{S} = \oint_C \mathbf{B} \mathrm{d} \mathbf{l} = \mu_0 \int_S \mathbf{J} \cdot \mathrm{d} \mathbf{S} \; .
\label{_auto12} \tag{12}
\end{equation}
$$

where we recognize that the curve $C$ is the curve limited by the surface $S$. We recall that $\int_S \mathbf{J} \cdot \mathrm{d} \mathbf{S}$ is the current $I$ passing through the surface $S$. We can therefore rewrite Ampere's law as:

<!-- Equation labels as ordinary links -->
<div id="_auto13"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d}\mathbf{l} = \mu_0 I_{\text{net}} \; .
\label{_auto13} \tag{13}
\end{equation}
$$

**Ampere's law.**

**Ampere's law on differential law** states that

<!-- Equation labels as ordinary links -->
<div id="_auto14"></div>

$$
\begin{equation}
\nabla \times \mathbf{B} = \mu_0 \mathbf{J} \; .
\label{_auto14} \tag{14}
\end{equation}
$$

**Ampere's law on integral form** states that:

<!-- Equation labels as ordinary links -->
<div id="_auto15"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = \mu_0 \int_S \mathbf{J} \cdot \mathrm{d} \mathbf{S} \; .
\label{_auto15} \tag{15}
\end{equation}
$$

where $C$ is a curve along the boundary of $S$.

**Ampere's law for a current loop** states that:

<!-- Equation labels as ordinary links -->
<div id="_auto16"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = \mu_0 I_{\text{net}} \; .
\label{_auto16} \tag{16}
\end{equation}
$$

where the current $I_{\text{net}}$ is the net current through a (any) surface that has the curve $C$ as its boundary.



### Let us address the various terms in Ampere's law in detail

1. The integral $\oint_C  \mathbf{B} \cdot d \, \mathbf{l}$ is a line integral of the magnetic field along a closed curve $C$. You have seen such integrals before. However, just as with Gauss' law, we will usually not perform this integral, but instead take the integral along a path where $\mathbf{B}$ is constant or where $\mathbf{B} \cdot d \, \mathbf{l}$ is constant. Hence, we can calculate the value of $B$ along the path.

2. The integral $\int_S \mathbf{J} \cdot d \, \mathbf{S}$ we recognize as the net current flowing through the surface $S$. The surface $S$ in Ampere's law is not an arbitrary surface. It must be a surface which has the curve $C$ as a boundary. In addition, the orientation of the surface $S$ must correspond to the orientation of the path $C$ according to the right-hand rule as illustrated in Fig. [fig:ampereslaw-curvesurface-010](#fig:ampereslaw-curvesurface-010). Notice that there are many surfaces $S$ that have $C$ as a boundary. We will therefore choose a surface that simplifies the calculation.

a. Instead of the integral $\int_S \mathbf{J} \cdot d \, \mathbf{S} = I$, we can use the current $I$. This current is the *net current* through the surface $S$. This means that we must add together all the currents going through the surface, taking into account the directions of the currents.

### Curve $C_1$

We see that for curve $C_1$ there are no currents passing through a planar surface through $C_1$. Therefore $I = 0$ and:

<!-- Equation labels as ordinary links -->
<div id="_auto17"></div>

$$
\begin{equation}
\oint_{C_1} \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = \mu_0 I = 0 \; .
\label{_auto17} \tag{17}
\end{equation}
$$

### Curve $C_2$

For curve $C_2$ there are two currents. First, we find the orientation of the planar surface $S_2$ with $C_2$ as the boundary. We see that current $I_1$ is going through a surface in the direction of the surface normal. Thus $I_1$ contributes positively to the total current through $C_2$. However, current $I_2$ has a direction that is opposite that of the surface normal to $S_2$. hence, $I_2$ contributes negatively to the total current. We find that:

<!-- Equation labels as ordinary links -->
<div id="_auto18"></div>

$$
\begin{equation}
\oint_{C_2} \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = \mu_0 \left( I_1 - I_2 \right) 
\label{_auto18} \tag{18}
\end{equation}
$$

### Curve $C_3$

For curve $C_3$ there are only one current that passes through the curve, $I_2$. We find the orientation of the surface $S_3$ and see that $I_2$ is in the same direction as the surface normal, hence the current $I_2$ contributes positively to the total current through $C_3$. This gives that:

<!-- Equation labels as ordinary links -->
<div id="_auto19"></div>

$$
\begin{equation}
\oint_{C_3} \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = \mu_0 I_2 
\label{_auto19} \tag{19}
\end{equation}
$$

## Example: Infinitely long cylidrical conductor

We want to find the magnetic field $\mathbf{B}$ around a cylindrical conductor using Ampere's law. The idea is to find an integration loop $C$ so that $\mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = B$ is a constant for this loop, $\oint_C \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = B \oint_C \mathrm{d} \, l = B L$, where $L$ is the length of the loop. We can then find $B$ from Ampere's law using $\oint_C \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = \mu_0 I$ by simply finding the current $I$ through a surface with $C$ as its boundary. However, to realize this plan we first need to find a symmetry for the magnetic field. The situation is illustrated in Fig. [fig:ampereslaw-inifitecylinder-010](#fig:ampereslaw-inifitecylinder-010).

### Symmetry of the current density

The magnetic field $\mathbf{B}$ is set up by the moving charges --- the currents. We therefore first describe the symmetry of the current density in the cylinder. The current density must be the same everywhere along the cylinder, because all points along the cylinder have *translational symmetry*. The current can therefore not flow out towards the edges, because the current would then flow out toward the edges everywhere along the cylinder and charges would pile up at the surface or move outside the cylinder. Because the cylinder also has rotational symmetry around the $z$-axis, we also assume that $\mathbf{J}$ has rotational symmetry. Thus, $\mathbf{J} = J(r) \mathbf{\hat{z}}$, where $r$ is the distance from the cylinder axis.

### Symmetry of the magnetic field from the symmetry of the cylinder

The *translational* symmetry of the infinite cylinder implies that the magnetic field cannot depend on $z$. The rotational symmetry around the cylinder implies that the magnetic field cannot depend on the azimuthal angle $\phi$. Thus, the field can only depend on the distance $r$ from the axis, $B = B(r)$. However, the translational and rotational symmetry still opens for the $\mathbf{B}$-field to have components in the $r$-, $\phi$- and $z$-direction: $\mathbf{B} = B_r(r) \mathbf{\hat{r}} + B_{\phi}(r) \boldsymbol{\hat{\phi}} + B_z(r) \mathbf{\hat{z}}$

### Symmetry of the magnetic field from Biot-Savart's law

We must use different types of physics arguments to say more about the direction of the magnetic field. We know that the magnetic field is set up by the currents. From Biot-Savart's law we know that the contribution from a volume element $dv$ is

<!-- Equation labels as ordinary links -->
<div id="_auto20"></div>

$$
\begin{equation}
d \, \mathbf{B} = \mu_0 \frac{\mathbf{J} dv \times \mathbf{R}}{4 \pi R^2} \; ,
\label{_auto20} \tag{20}
\end{equation}
$$

where the vector $\mathbf{R}$ points from the element $dv$ to the observation point. We argued above that $\mathbf{J} = J \mathbf{\hat{z}}$. This means that there will not be any component along $z$ from $d \, \mathbf{B}$ due to the cross product (the cross product is normal to the $z$-axis). Therefore $B_z(r) = 0$.

### Symmetry of the magnetic field from zero-divergence condition

Can the magnetic field have a component in the $r$-direction, $B_r(r)$? No, because this would imply that there is a net flux of the magnetic field out of a cylindrical surface. If we place a cylinderical (Gauss) surface around the $z$-axis, there would be no flux of the magnetic field out of the side surfaces (because $B_z = 0$). The flux out of the surface would then be $B_r(r) 2 \pi r L$. However, this must be zero because $\oint_S \mathbf{B} \cdot d \, \mathbf{S} = 0$. Therefore, $B_r(r) = 0$.

### The simplified magnetic field

The argument presented here was long and systematic. When you get used to this you should be able to make these types of considerations quickly without all the details of the argument. We have now found that the magnetic field must have the form

<!-- Equation labels as ordinary links -->
<div id="_auto21"></div>

$$
\begin{equation}
\mathbf{B} = B_{\phi}(r) \boldsymbol{\hat{\phi}} \; .
\label{_auto21} \tag{21}
\end{equation}
$$

### Application of Ampere's law to find the magnetic field

We are now ready to use Ampere's law to find the magnetic field. Because the field only depends on $r$ we should choose a curve with a constant $r$ so that the magnitude of the field is constant. We should also choose a curve so that the magntic field and the tangential vector $d \, \mathbf{l}$ along the curve always has the same relative angle. We do this by choosing a circular curve in the $xy$-plane with its center at the $z$-axis. If we choose the curve to be oriented in the positive rotational direction around the $z$-axis (according to the right-hand-rule), then $d \, \mathbf{l} = dl \boldsymbol{\hat{\phi}}$ and

<!-- Equation labels as ordinary links -->
<div id="_auto22"></div>

$$
\begin{equation}
\oint_c \mathbf{B} \cdot d \, \mathbf{l} = \oint_C B_{\phi}(r) d \, l = B 2 \pi r = \mu_0 I(r) \; .
\label{_auto22} \tag{22}
\end{equation}
$$

Where $I$ is the current going through a cross section of the cylinder (inside the radius $r$). If $r>a$ then all the current $I$ goes through the curve and $I(r) = I$ is the the total current in the cylinder. The magnetic field is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto23"></div>

$$
\begin{equation}
\mathbf{B} = B_{\phi} \boldsymbol{\hat{\phi}} = \frac{\mu_0 I}{2 \pi r} \boldsymbol{\hat{\phi}}  \; .
\label{_auto23} \tag{23}
\end{equation}
$$

### What if $r<a$ ?

There is nothing in our arguments that implies that $r$ must be larger than $a$. However, if $r<a$ we must ensure that $I(r)$ only includes the part of the current that is inside the loop of radius $r$. If we assume that the current density $\mathbf{J}$ is uniform throughout the cylinder we can calculate the current $I(r)$:

<!-- Equation labels as ordinary links -->
<div id="_auto24"></div>

$$
\begin{equation}
I(r) = \oint_S \mathbf{J}\cdot d \, \mathbf{S} = J \pi r^2 \; .
\label{_auto24} \tag{24}
\end{equation}
$$

But what is $J$? We can find this by setting $r = a$, for which $I(a) = I = J \pi a^2$, hence $J = I/(\pi a^2$, and

<!-- Equation labels as ordinary links -->
<div id="_auto25"></div>

$$
\begin{equation}
I(r) = \frac{I}{\pi a^2} \pi r^2 = I \frac{r^2}{a^2} \; .
\label{_auto25} \tag{25}
\end{equation}
$$

The magnetic field is then for $r < a$:

<!-- Equation labels as ordinary links -->
<div id="_auto26"></div>

$$
\begin{equation}
\mathbf{B} = \frac{\mu_0 I(r)}{2 \pi r} \boldsymbol{\hat{\phi}} = \frac{\mu_0 I r}{2 \pi a^2} \boldsymbol{\hat{\phi}} \; .
\label{_auto26} \tag{26}
\end{equation}
$$

The complete field is then

<!-- Equation labels as ordinary links -->
<div id="_auto27"></div>

$$
\begin{equation}
\mathbf{B} = \left\{ \begin{array}{cc}
\frac{\mu_0 I}{2 \pi r} \boldsymbol{\hat{\phi}} & \, , \, r \ge a \\ 
\frac{\mu_0 I r}{2 \pi a^2} \phi & \, , \, r<a
\end{array} \right.
\label{_auto27} \tag{27}
\end{equation}
$$

### Typical numbers for the magnetic field

If a current of $I = 10 \text{A}$ is carried by the wire (the cylindrical conductor), what is the magnetic field at a distance of $r = 10 \text{cm}$ from the wire? We find that

<!-- Equation labels as ordinary links -->
<div id="_auto28"></div>

$$
\begin{equation}
B = \frac{4 \pi \, 10^{-7} \, 10 }{2 \pi \, 0.1} \text{T} = 2 \cdot 10^{-5} \text{T} = 20 \mu \text{T} \; .
\label{_auto28} \tag{28}
\end{equation}
$$

This is an typical magnitude of a magnetic field from the wires in your house.


**Method: Applying Ampere's law to find the magnetic field.**


Based on this example, we propose a general strategy for how to find the magnetic field for a current distribution (from a current loop) using Ampere's law:

* Find a (integration) loop $C$ the bound a surface so that $\mathbf{B} \cdot \mathrm{d} \mathbf{l} = B$ is a constant on this loop. (It may be zero on some parts of the loop --- zero is a constant. And remember that it is only the component of $\mathbf{B}$ along $\mathrm{d} \mathbf{l}$ that needs to be constant. Other elements are ignored in this approach. We call this loop the *Ampere loop* for the system.

* This usually requires that you find a simplified description of the magnetic field in a coordinate system that has the symmetry of the system, e.g. cylindrical coordinates for a cylindrical system we often find that $\mathbf{B} = B_\phi (r) \boldsymbol{\hat{\phi}}$.

* You often also have to use that the divergence of $\mathbf{B}$ is zero to argue for the symmetry of the $\mathbf{B}$-field.

* Find the curve integral, which often can be simplified to only include the length of the curve: $\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = B_l \oint_C \mathrm{d} l$

* Use Ampere's law $\mu_0 I = B \oint_C \mathrm{d} l$, to find the magnetic field as a function of current and position.

* Notice that the curve sometimes will enclose the current several times, for example when the wire is twisted several times around the magnetic field as in a solenoid. You must then include the current through all the loops, or multiply the current by the number of loops.



## Example: Coaxial cable



A coaxial cable consists of a conducting (inifinitely thin) sheet of radius $a$ and an (infinitely) thin outer shell at a radius $b$. Find the magnetic field when a current $I$ is flowing in one direction in the center and in the opposite direction in the outer sheet. (Alternatively, we can assue that the conductors have a finite thickness, but that all the current is on the surface of the conductor).

We can use the same symmetry argument as for the inifinitely long cylinder, therefore $\mathbf{B} = B_{\phi}(r) \boldsymbol{\hat{\phi}}$ and we are left with Ampere's law

<!-- Equation labels as ordinary links -->
<div id="_auto29"></div>

$$
\begin{equation}
\oint_c \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = B 2 \pi r = \mu_0 I_{\text{net}} \; .
\label{_auto29} \tag{29}
\end{equation}
$$

When $r<a$ there is no current inside the loop, hence $B=0$ in this region. When $r>b$ the *net* current is zero, hence $B = 0$ in this region as well. In the intermediate region we have that the net current is $I$. The magnetic field is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto30"></div>

$$
\begin{equation}
\mathbf{B} = \left\{
\begin{array}{cc}
0 & \, , \, r<a \\ 
\frac{\mu_0 I}{2 \pi r} \boldsymbol{\hat{\phi}} & \, , \, a \le r \le b \\ 
0 & \, , \, b<r
\end{array} \right.
\label{_auto30} \tag{30}
\end{equation}
$$

## Example: Toroid
There are several ways to increase the current through a surface: We can increase the current carried in a single wire or we can place many wires through the same surface. There are two main geometries where this is done systematically to generate a well-controlled magnetic field in a region of space: the *toroid* and the *solenoid*. First, let us address the behavior of the toroid.

### Toroid geometry

The toroid consists of a donut shaped material with a wire wound around it as illustrated in Fig. [fig:ampereslaw-toroid-010](#fig:ampereslaw-toroid-010). The wire connects to the outside on one side and is usually wound tightly. Typically we assume that it is wound so tightly that we can assume that the wire consists of circles that each are in the plane of the radial direction (radial from the center of the toroid) and the $z$-axis.

<!-- dom:FIGURE:[fig-ampereslaw/ampereslaw-toroid-010.png, height=400 width=600 frac=1.0] Illustration of the toroid geometry. <div id="fig:ampereslaw-toroid-010"></div> -->

<div id="fig:ampereslaw-toroid-010"></div>

<p>Illustration of the toroid geometry.</p>
<img src="fig-ampereslaw/ampereslaw-toroid-010.png" height=400 width=600>

<!-- end figure -->


### Symmetry

What is the expected symmetry of the $\mathbf{B}$-field in this case? We expect it to depend on the distance $r$ from the center of the toroid. We do not expect it to depend on the angular position $\phi$. We do not expect any field in the $r$-direction, since the current is in this direction. We also do not expect any $z$-component for the field since there is no current through the loop illustrated in Fig. [fig:ampereslaw-toroid-010](#fig:ampereslaw-toroid-010)d, and therefore $\oint_{C_3} \mathbf{B}\cdot d \, \mathbf{l} = 0$. We place the outside loop very far away and are then left with the $B_z$ component close to the toroid, which must be zero. The symmetry of $\mathbf{B}$ is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto31"></div>

$$
\begin{equation}
\mathbf{B} = B(r) \boldsymbol{\hat{\phi}} \; .
\label{_auto31} \tag{31}
\end{equation}
$$

### Applying Ampere's law

We can now apply Ampere's law along a path where $\mathbf{B} \cdot d \, \mathbf{l}$ is constant, that is, for a circle with radius $r$ as illustrated in Fig. [fig:ampereslaw-toroid-010](#fig:ampereslaw-toroid-010)c. For this path we have:

<!-- Equation labels as ordinary links -->
<div id="_auto32"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = B 2 \pi r = I \mu_0 = - N I \mu_0 \; ,
\label{_auto32} \tag{32}
\end{equation}
$$

where we have used that the total current through the loop $C$ is the sum of the currents in all the $N$ windings of the wire around the toroid. This result is only valid inside the toroid. Outside the toroid there is not net current through the surface and the field is zero. This means that the field is:

<!-- Equation labels as ordinary links -->
<div id="_auto33"></div>

$$
\begin{equation}
\mathbf{B} = \left\{ \begin{array}{cc}
- \frac{\mu_0 N I }{2 \pi r} \boldsymbol{\hat{\phi}} & \text{ inside the toroid } \\ 
0 & \text{ outside }
\end{array}
\right.
\label{_auto33} \tag{33}
\end{equation}
$$

A solenoid is what we get if we wind a wire around a long cylinder. The geometry is illustrated in Fig. [fig:ampereslaw-solenoid-010](#fig:ampereslaw-solenoid-010). Notice how we use the standard symbols for vectors coming out of/going into the page to illustrate the direction of the currents. We assume that the wire has been wound $N$ times along a length $L$ of the cylinder, with $N/L$ windings per unit length.

### Solenoid and Toroid

We could assume that the solenoid is what we get for a toroid in the limit when the radius of the toroid becomes very huge. The result would therefore be a constant magnetic field inside the solenoid and zero field outside the solenoid. For the toroid the number $N/L$ of windings per unit length is $N/L = N/2 \pi r$, so that the field is $\mathbf{B} = \mu_0 I (N/L) \mathbf{\hat{z}}$, where $z$ is the axis long the solenoid.

### Direct calculation of the solenoid

Let us instead do a direct calculation for the solenoid starting from the symmetry of the magnetic field in the solenoid and then applying Ampere's law for a well-chosen curve.

### Symmetry of the solenoid

First, we look at the $\phi$-component of the field. We can use Amperes law for a circulat path around the cylinder. There is no net current through this curve $C$ (see Fig. [fig:ampereslaw-solenoid-010](#fig:ampereslaw-solenoid-010)). Thus, Ampere's law gives:

<!-- Equation labels as ordinary links -->
<div id="_auto34"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \, \mathbf{l} = \oint_C B_{\phi} \mathrm{d} \, l = B_{\phi} 2 \pi r = 0 \, \Rightarrow \, B_{\phi} = 0 \; .
\label{_auto34} \tag{34}
\end{equation}
$$

For the $r$-component in the direction radially out from the $z$-axis, we can apply the divergence theorem for the $\mathbf{B}$-field. where

<!-- Equation labels as ordinary links -->
<div id="_auto35"></div>

$$
\begin{equation}
\int \mathbf{B} \cdot \mathrm{d} \, \mathbf{S} = B_r 2 \pi r L  = 0 \, \Rightarrow \, B_r = 0 \; .
\label{_auto35} \tag{35}
\end{equation}
$$

For the $z$-component we notive that $B_z(r)$ cannot depend on the the $z$ or $\phi$ coordinates due to the symmetry: We assume that the cylinder is long and that edge effects are small. $B_z$ therefore cannot depend on $z$ cause the system has translational symmetry along the $z$-axis. Similarly, $B_z$ cannot depend on $\phi$ because the system has rotational symmetry. We are therefore left with $\mathbf{B} = B_z(r) \mathbf{\hat{z}}$. Outside the cylinder the field must be zero. We place a circuit with one part of length $l$ inside the cylinder (see Fig. [fig:ampereslaw-solenoid-010](#fig:ampereslaw-solenoid-010)d) and one part of length $l$ outside the cylinder. The integral in Ampere's law is thefore:

<!-- Equation labels as ordinary links -->
<div id="_auto36"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \, \mathrm{d} \mathbf{l} = Bl = \mu_0 I_{\text{net}} \; .
\label{_auto36} \tag{36}
\end{equation}
$$

In this case, $I_{\text{net}} = (N/L)l I$, where $N/L$ is the nuber windings per unit length so that $(N/L)l$ is the number of windings inside the curve of length $l$. The field is therefore

<!-- Equation labels as ordinary links -->
<div id="_auto37"></div>

$$
\begin{equation}
\mathbf{B} = B_z \mathbf{\hat{z}} = \mu_0 I \frac{N}{L} \; , \mathbf{\hat{z}} \; .
\label{_auto37} \tag{37}
\end{equation}
$$

The magnetic field is therefore uniform inside the solenoid. The solenoid is therefore the magnetic equivalent of the capacitor --- a simple geometry where the field is uniform inside it.

# Comparison of magnetostatics and electrostatics

We have now discovered the main laws of electro- and magneto-*statics*, and we have found that they have a nice kind of symmetry. For electrostatics we found that

$$
\begin{eqnarray}
\nabla \cdot \mathbf{E} &=& \frac{1}{\epsilon_0} \rho \quad \text{ (Gauss' law)} \\ 
\nabla \times \mathbf{E}&=& 0 
\end{eqnarray}
$$

These two laws are called **Maxwell's equations** for electrostatics. We can use these to find the electric field for any charge distribution. It is only the second equation that will be modified (with Faraday's law) when we introduce a time-varying magnetic field. These equations can be found from Coulomb's law, and we can show that they reproduce Coulomb's law.

For magnetostatics we have that

$$
\begin{eqnarray}
\nabla \cdot \mathbf{B} &=& 0 \\ 
\nabla \times \mathbf{B}&=& \mu_0 \mathbf{J} \quad \text{ (Ampere's law) } 
\end{eqnarray}
$$

These two laws are called **Maxwell's equations** for magnetostatics. Together with Maxwell's equations for electrostatics, they form the four Maxwell's equations. It is only the second equation that will be modified when we introduce time-varying electric fields. We can use these two equations to find the magnetic field for a given distribution of currents.

In addition to Maxwell's equations, we have Lorentz' law, which show how the fields interacts with charges:. The force on a charge $Q$ is:

<!-- Equation labels as ordinary links -->
<div id="_auto38"></div>

$$
\begin{equation}
\mathbf{F} = Q \left( \mathbf{E} + \mathbf{v} \times \mathbf{B} \right) \; .
\label{_auto38} \tag{38}
\end{equation}
$$

These laws essentially provides the entire description of (static) electromagnetism. We will modify them as we introduce dynamics, that is, time-varying fields. It is simple and beautiful!

# Proof of Ampere's law

To prove Ampere's law we will use the vector potential representation of the magnetic field, where $\mathbf{B} = \nabla \times \mathbf{A}$ and

<!-- Equation labels as ordinary links -->
<div id="_auto39"></div>

$$
\begin{equation}
\mathbf{A} = \frac{\mu_0}{4 \pi} \int \frac{\mathbf{J}}{R}\mathrm{d} v' \; ,
\label{_auto39} \tag{39}
\end{equation}
$$

where the integral is over $\mathbf{r}'$. Ampere's law states that $\nabla \times \mathbf{B} = \mu_0 \mathbf{J}$. Let us calculate $\nabla \times \mathbf{B}$ using vector relations. We know that the general vector relation

<!-- Equation labels as ordinary links -->
<div id="_auto40"></div>

$$
\begin{equation}
\mathbf{a} \times \left( \mathbf{b} \times \mathbf{c} \right) = \mathbf{b} \left( \mathbf{a}\cdot \mathbf{c} \right) - \mathbf{c} \left( \mathbf{a} \cdot \mathbf{b}\right) \; ,
\label{_auto40} \tag{40}
\end{equation}
$$

always is true. We use this to find $\nabla \times \left( \nabla \times \mathbf{A}\right)$, getting:

<!-- Equation labels as ordinary links -->
<div id="eq:ampereslaw-proof-110"></div>

$$
\begin{equation}
\nabla \times \left( \nabla \times \mathbf{A}\right) = \nabla \left( \nabla \cdot \mathbf{A} \right) - \left( \nabla \cdot \nabla \right) \mathbf{A} = \nabla \left( \nabla \cdot \mathbf{A}\right) - \nabla^2 \mathbf{A} \; .
\label{eq:ampereslaw-proof-110} \tag{41}
\end{equation}
$$

Let us start with the second term, $\nabla^2 \mathbf{A}$. This is the Laplacian of a vector function $\mathbf{A}$. In Cartesian coordinates this corresponds to the Laplacian for each of coordinates:

<!-- Equation labels as ordinary links -->
<div id="_auto41"></div>

$$
\begin{equation}
\nabla^2\mathbf{A} = \nabla^2 A_x \mathbf{\hat{x}} + \nabla^2 A_y \mathbf{\hat{y}} + \nabla^2 A_z \mathbf{\hat{z}} \; ,
\label{_auto41} \tag{42}
\end{equation}
$$

Where

<!-- Equation labels as ordinary links -->
<div id="_auto42"></div>

$$
\begin{equation}
A_x =  \frac{\mu_0}{4 \pi} \int \frac{J_x}{R}\mathrm{d} v'
\label{_auto42} \tag{43}
\end{equation}
$$

so that

$$
\begin{equation}
\nabla^2 A_x = \frac{\mu_0}{4 \pi} \int \nabla^2 \frac{J_x}{R}\mathrm{d} v'
\label{}
\end{equation}
$$

Now, we will use a trick, where we use our experience from electrostatics to solve this equation. From electrostatics we remember that

<!-- Equation labels as ordinary links -->
<div id="_auto43"></div>

$$
\begin{equation}
V = \frac{1}{4 \pi \epsilon_0} \int \frac{\rho \mathrm{d} v'}{R} 
\label{_auto43} \tag{44}
\end{equation}
$$

and Poisson's equation $\nabla^2 V = - \rho/\epsilon_0$, that is,

<!-- Equation labels as ordinary links -->
<div id="_auto44"></div>

$$
\begin{equation}
\nabla^2 V = \frac{1}{4 \pi \epsilon_0} \int \nabla^2 \frac{\rho}{R}\mathrm{d} v' = - \frac{\rho}{\epsilon_0}
\label{_auto44} \tag{45}
\end{equation}
$$

and

<!-- Equation labels as ordinary links -->
<div id="_auto45"></div>

$$
\begin{equation}
\frac{1}{4 \pi} \int \nabla^2 \frac{\rho}{R}\mathrm{d} v' = - \rho 
\label{_auto45} \tag{46}
\end{equation}
$$

This must be true for any $\rho$. For example, if we instead choose $\rho$ to be $J_x$. That is, we have demonstrated that

<!-- Equation labels as ordinary links -->
<div id="_auto46"></div>

$$
\begin{equation}
\frac{1}{4 \pi} \int \nabla^2 \frac{J_x}{R}\mathrm{d} v' = - J_x \;.
\label{_auto46} \tag{47}
\end{equation}
$$

The same is true for the other component of $\mathbf{J}$. We therefore get that

<!-- Equation labels as ordinary links -->
<div id="_auto47"></div>

$$
\begin{equation}
\nabla^2 \mathbf{A} = - \mu_0 \mathbf{J} 
\label{_auto47} \tag{48}
\end{equation}
$$

This is the *vector Poisson's equation* and can be used to find the magnetic field given the current distributions in the same way as we used Poisson's equation to find the electric field from the charge distribution. However, the equation is somewhat more complicated because it is a vector equation, but we can in principle use many of the same tools as we have developed for Laplace's and Poisson's equation.

Now, we have found one of the two terms in ([41](#eq:ampereslaw-proof-110)). What about the first term, $\nabla \left( \nabla \cdot \mathbf{A}\right)$? Let us look at $\nabla \cdot \mathbf{A}$ in detail. It is:

<!-- Equation labels as ordinary links -->
<div id="_auto48"></div>

$$
\begin{equation}
\nabla \cdot \mathbf{A} = \frac{\mu_0}{4 \pi}\nabla \cdot \int \frac{\mathbf{J}(\mathbf{r}')\mathrm{d} v'}{R} = \frac{\mu_0}{4 \pi} \int \nabla \cdot \left( \frac{\mathbf{J}}{R}\right) \mathrm{d} v' \; ,
\label{_auto48} \tag{49}
\end{equation}
$$

where we have used that we can take $\nabla$ inside the integral, because the integral is over $\mathbf{r}'$ and $\nabla$ is with respect to $\mathbf{r}$, which is a constant during the integration. We can now use the divergence theorem to rewrite this integral:

<!-- Equation labels as ordinary links -->
<div id="_auto49"></div>

$$
\begin{equation}
\frac{\mu_0}{4 \pi} \int \nabla \cdot \left( \frac{\mathbf{J}}{R}\right) \mathrm{d} v' =
\frac{\mu_0}{4 \pi }\oint_S \frac{\mathbf{J}}{R} \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto49} \tag{50}
\end{equation}
$$

where the surface $S$ is enclosing volume $v$. If we let the volume $v$ increase and hence the surface $S$ become larger and larger, all the currents will eventually only be inside the volume and no currents will cross the surface $S$ and the integral will approach zero. We therefore conclude that $\nabla \cdot \mathbf{A} = 0$.

We have therefore now found that

<!-- Equation labels as ordinary links -->
<div id="_auto50"></div>

$$
\begin{equation}
\nabla \left( \nabla \times \mathbf{A} \right) = \mu_0 \mathbf{J} \; ,
\label{_auto50} \tag{51}
\end{equation}
$$

and therefore that

<!-- Equation labels as ordinary links -->
<div id="_auto51"></div>

$$
\begin{equation}
\nabla \times \mathbf{B} = \mu_0 \mathbf{J} \; ,
\label{_auto51} \tag{52}
\end{equation}
$$

and we have proved Ampere's law.


# Summary

The **magnetic flux** through a closed surface is zeroa charge $Q_2$ in $\mathbf{r}_2$ with velocity $\mathbf{v}_2$ is

$$
\oint_A \mathbf{B} \cdot \mathrm{d} \mathbf{S} = 0 \; ,
$$

and the **divergence of the magnetic field is zero**

$$
\nabla \cdot \mathbf{B} = 0
$$

**Ampere's law on differential form** states that

$$
\nabla \times \mathbf{B} = \mu_0 \mathbf{J} \; .
$$

**Amperes's law on integral form** states that

$$
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = \mu_0 \int_S \mathbf{J} \cdot \mathrm{d} \mathbf{S} \; ,
$$

and **Ampere's law for a current loop** states that

$$
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = \mu_0 I_{\text{net}} \; .
$$

We find the magnetic field using Amperes law by:
* Finding an integration loop $C$ where $\mathbf{B}\cdot \mathrm{d} \mathbf{l}$ is a constant due to symmetry

* Apply Ampere's law to find the magnetic field as a function of current and position.

There exists a **vector potential** $\mathbf{A}$ so that $\mathbf{B} = \nabla \times \mathbf{A}$. We can find the vector potential from

$$
\mathbf{A} = \frac{\mu_0}{4  \pi}\int_v \frac{\mathbf{R}\mathrm{d} v'}{R} \; ,
$$

or from the **vector Poisson's equation**:

$$
\nabla^2 \mathbf{A} = - \mu_0 \mathbf{J}
$$

The **divergence of the vector potential is zero**.

$$
\nabla \cdot \mathbf{A} = 0 \; .
$$
