We have now introduced the basic law of magnetic induction,  Faraday's law, which demonstrates how a time-varying magnetic field induces a emf in a circuit. For any circuit carrying a time-varying current, we know that the current will generate a time-varying magnetic field, and the magnetic field will generate a time-varying flux through the circuit, which again will induce an emf in the circuit. In this chapter, we will model this effect in a circuit by introducing the self-inductance of the circuit. The self-inductance is defined as the flux per unit current in a way similar to how we introduced capacitance as the charge per potential difference or the resistance as the potential difference per unit current. We will demonstrate how we can introduce a new element in a circuit, an inductor, to represent the inductance of the circuit, and we will show how you can calculate the inductance of a circuit based on the geometry of the current-carrying components and the magnetic materials. We will extend the concept of inductance to also include how one circuit can induce an emf in another circuit --- through the mutual inductance of the two circuits, which again only depends on the geometry of the system. Finally, we will introduce the energy of the magnetic field in a capacitor in a way similar to how we introduced the energy of the electric field for a capacitor. 

  ======= Inductance =======

Fig. [fig:inductance-circuit-010](#fig:inductance-circuit-010) illustrates a circuit made of a conductor of finite thickness. If a current runs in the circuit, a magnetic field will be generated, and the magnetic field will have a flux through the circuit itself. This self-induced flux is called a *self-inductance* and we introduce the term $L$ to describe this effect a circuit has on itself:

<!-- Equation labels as ordinary links -->
<div id="_auto1"></div>

$$
\begin{equation}
\Phi = L \, I \quad , \quad L = \frac{\Phi}{I} \; .
\label{_auto1} \tag{1}
\end{equation}
$$

When we calculate the flux $\Phi$ in this expression, it is only the flux from the magnetic field generated by the circuit itself that is to be included:

<!-- Equation labels as ordinary links -->
<div id="_auto2"></div>

$$
\begin{equation}
\Phi = \int_S \mathbf{B}\cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto2} \tag{2}
\end{equation}
$$

where $\mathbf{B}$ is generated by $I$. There may be other sources of magnetic fields, such as a permanent magnetic or another coil nearby, but these are not included in the flux $\Phi$ which is included in the self-inductance.

## Definition of self-inductance

**Self-inductance.**

The **self-inductance** $L$ of a system/circuit is defined as

<!-- Equation labels as ordinary links -->
<div id="_auto3"></div>

$$
\begin{equation}
L = \frac{\Phi}{I} \; ,
\label{_auto3} \tag{3}
\end{equation}
$$

where the flux $\Phi$ is the flux through the circuit from the magnetic field generated by the current $I$.



## What determines the self-inductance $L$?

For a circuit as illustrated in Fig. [fig:inductance-circuit-010](#fig:inductance-circuit-010), the magnetic field $\mathbf{B}$ is generated by the current as described by Biot-Savart's law:

<!-- Equation labels as ordinary links -->
<div id="_auto4"></div>

$$
\begin{equation}
\mathbf{B} = \int_C \frac{\mu_0 I \mathrm{d} \mathbf{l}\times \mathbf{\hat{R}} }{4 \pi R^2} \; ,
\label{_auto4} \tag{4}
\end{equation}
$$

and the flux is then

<!-- Equation labels as ordinary links -->
<div id="_auto5"></div>

$$
\begin{equation}
\Phi = \int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto5} \tag{5}
\end{equation}
$$

integrated over a surface $S$ enclosed by the circuit/path $C$. This shows that

* The magnetic field is proportional to the current and the flux is proportional to the magnetic field, $\Phi \propto B \propto I$. Thus there is a *linear* relationship between $\Phi$ and $I$. This motivates the definition of $\Phi = LI$ so that $L$ describes this constant of proportionality.

* The inductance $L$ is then determined by the geometry of the system. It is the geometry of the circuit and the presence of magnetic materials that determine $\mathbf{B}$. And it is then again the geometry of the circuit that determines the flux.

* The definition of $L$ is therefore very similar to our previous definitions of the capacitance $C = Q/V$ or the resistance $R = V/I$ of a system.

* This also suggests a *method to find the inductance*: We assume a given $I$, calculate the resulting $\mathbf{B}$ and the resulting flux $\Phi$ and find $L$ from $L = \Phi/I$. This method is effectively identical to the method we introduced for the capacitance and the resistance.

**Method: Calculate the inductance.**

* Assume that a current $I$ runs through the circuit. (Often you will only have to look at the part of the circuit that acts as a coil with many windings).

* Calculate the magnetic field $\mathbf{B}$ from the current $I$ using Biot-Savart's law or Ampere's law. Include effects of magnetic materials.

* Calculate the flux of the magnetic field $\mathbf{B}$ set up by the current $I$ through a convenient surface $S$ enclosed by the circuit path $C$

* Find the self-inductance $L$ from $L = \Phi/I$.



## How does the self-inductance enter the circuit equations?

The self-inductance of a circuit describes the relation between the flux $\Phi$ and the current $I$ in a circuit, $\Phi = L I$. But how does this affect the behavior of the circuit? If the current is *time-varying*, then the current will induce a time-varying flux:

<!-- Equation labels as ordinary links -->
<div id="_auto6"></div>

$$
\begin{equation}
I = I(t) \quad \Rightarrow \quad \Phi(t) = L I = L I(t) \; .
\label{_auto6} \tag{6}
\end{equation}
$$

This will again introduce an emf to the circuit according to Faraday's law:

<!-- Equation labels as ordinary links -->
<div id="_auto7"></div>

$$
\begin{equation}
e = - \frac{\mathrm{d} }{\mathrm{d} t} \Phi(t) = - \frac{\mathrm{d} }{\mathrm{d} t} L I(t) \; .
\label{_auto7} \tag{7}
\end{equation}
$$

If the geometry of the circuit does not change, then $\mathrm{d} L/ \mathrm{d} t = 0$ and we get:

<!-- Equation labels as ordinary links -->
<div id="_auto8"></div>

$$
\begin{equation}
e = - L \frac{\mathrm{d} I}{\mathrm{d} t} \; .
\label{_auto8} \tag{8}
\end{equation}
$$

To include this effect in a circuit, we introduce an element $L$ in the circuit as illustrated in Fig. [fig:inductance-circuit-010](#fig:inductance-circuit-010)b. This element is drawn like a coil in a circuit diagram.

This element is in principle a part of any circuit. For the circuit illustrated in Fig. [fig:inductance-circuit-010](#fig:inductance-circuit-010), we may introduce an inductance $L$ to model the effect of the geometry of the circuit. We must also include the effect of the emf in Kirchoff's voltage law for the circuit. We include the emf $e$ in addition to the other voltage sources $e_j$ in the circuit:

<!-- Equation labels as ordinary links -->
<div id="_auto9"></div>

$$
\begin{equation}
\sum e_j - \sum_i V_i = 0 \; ,
\label{_auto9} \tag{9}
\end{equation}
$$

where the $V_i$ represents the voltage drops, such as $V_1 = RI$ for a resistor, whereas the emf's represents batteries, $V_b$ and the emf due to inductance, $e = - \mathrm{d} \Phi / \mathrm{d} t$. For the illustrated circuit we have the emf $V_b + e$ and

<!-- Equation labels as ordinary links -->
<div id="_auto10"></div>

$$
\begin{equation}
V_b + e - RI = 0 \; .
\label{_auto10} \tag{10}
\end{equation}
$$

All circuits have some inductance, often as an undesirable side effect. We represent this inductance by a coil (inductor) circuit element. However, we also design and include inductors that play important roles of circuits, such as a coil. We call such circuit elements *inductors*. Inductors can store magnetic energy, and play an important role in many circuits. In the next chapter, we will address how we construct circuits using the basic circuit elements resistor, capacitor and inductor, and how we can use circuits as models of model complex, real systems.

Notice that the model we have introduced for an inductor is a simplified model of the processes that goes on in the electromagnetic system, where we usually assume that the whole circuit can be described as a set of independent components. This is an approximation that sometimes breaks down. Then we have to apply the full machinery of an electromagnetic analysis.

## Effect of self-inductance in a circuit


What is the effect of the self-inductance $L$ in the circuit illustrated in Fig. [fig:inductance-circuit-010](#fig:inductance-circuit-010)? In this circuit, we found that

<!-- Equation labels as ordinary links -->
<div id="_auto11"></div>

$$
\begin{equation}
V_b + e - RI = V_b - L\frac{\mathrm{d}}{\mathrm{d} t} I - RI = 0 \; .
\label{_auto11} \tag{11}
\end{equation}
$$

What happens if a switch in the circuit is suddenly turned on? In this case, we expect the current to increase rapidly. The effect of the self-inducance is to induce an emf $e = - L \mathrm{d} I / \mathrm{d} t$. If the current increases rapidly, this will generate a large emf, which will counteract the change: If the current is increasing, the emf will be negative, whereas if the current is decreasing, the emf will be positive. The self-inductance will therefore act to reduce changes in current.

Similarly, if we very rapidly turn off a circuit, for example by unplugging a wire, there will be a very large emf due to the rapid change in current. This is what induces sparks when you unplug an electric device. We will address the effect of inductance in circuits in the next chapter. In this chapter, we will focus on how to calculate the inductance for a given circuit geometry, just like we did for capacitors and resistors.

## Example: Toroid


Fig. [fig:inductance-toroid-010](#fig:inductance-toroid-010) illustrates a circuit wrapped $N$ times around a toroid. The toroid has an inner radius $a$ and an outer radius $b$, and a rectuangular cross section of height $h$. Find the self-inductance $L$ of this circuit.

<!-- begin figure -->
<div id="fig:inductance-toroid-010"></div>

<p>Illustration of a toroid with rectangular cross section, inner radius $a$, outer radius $b$, heigh $h$ and $N$ windings.</p>

<!-- end figure -->


**Solution:**

Our plan is the use the method described above: We will assume a current $I$ in the wire, find the resulting magnetic field $\mathbf{B}$ in the region enclosed by the circuit, calculate the resulting flux $\Phi$ and find the (self) inductance from $L = \Phi/I$.

### Finding the magnetic field

We find the magnetic field using Ampere's law. First, we realize from symmetry and that the divergence of the magnetic field is zero that the magnetic field only has an azimuthal component, which only can depend on the distance $r$ to the center of the toroid:

<!-- Equation labels as ordinary links -->
<div id="_auto12"></div>

$$
\begin{equation}
\mathbf{B} = B_{\phi}(r) \boldsymbol{\hat{\phi}} = B(r) \boldsymbol{\hat{\phi}} \; .
\label{_auto12} \tag{12}
\end{equation}
$$

We apply Ampere's law on a circular loop $C$ or radius $r$ inside the toroid. The magnetic field is constant along this circle is constant. Ampere's law gives that:

<!-- Equation labels as ordinary links -->
<div id="_auto13"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = B \, 2 \pi r = \mu_0 I_{\text{tot}} \; .
\label{_auto13} \tag{13}
\end{equation}
$$

The total current going through the circle is $I$ times the number of times the wire intersects the circle, which is $N$ times, thus $I_{\text{tot}} = N I$. We find that

<!-- Equation labels as ordinary links -->
<div id="_auto14"></div>

$$
\begin{equation}
\mathbf{B} = \frac{\mu_0 N I}{2 \pi r} \boldsymbol{\hat{\phi}} \; .
\label{_auto14} \tag{14}
\end{equation}
$$

### Finding the flux

The flux through a single turn of the wire is

<!-- Equation labels as ordinary links -->
<div id="_auto15"></div>

$$
\begin{equation}
\Phi_1 = \int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto15} \tag{15}
\end{equation}
$$

where the integration is over a cross-sectional area from $r = a$ to $r = b$. Notice that $\mathbf{B}$ is not uniform across this area. It points in the same direction across the whole area, in a direction normal to the wire loop, but the magntide varies with $r$. A small surface element is $\mathrm{d} S = h \mathrm{d} r$. The flux of a single turn is therefore:

$$
\begin{eqnarray}
\Phi_1 &=& \int_a^b B h \mathrm{d} r = \int_a^b \frac{\mu_0 N I }{2 \pi r} h \mathrm{d} r \\ 
&=& h \frac{\mu_0 N I}{2 \pi} = \int_a^b \frac{1}{r}\mathrm{d} r \\ 
&=& \frac{h \mu_0 N I}{2 \pi} \ln \frac{b}{a} \; .
\end{eqnarray}
$$

This is the flux through a single winding. There are $N$ windings, so that the total flux is:

<!-- Equation labels as ordinary links -->
<div id="_auto16"></div>

$$
\begin{equation}
\Phi_N = N \Phi_N = \frac{h \mu_0 N^2 I}{2 \pi} \ln \frac{b}{a} \; .
\label{_auto16} \tag{16}
\end{equation}
$$

Notice that the number of windings enters this expression twice: Once in the magnetic field and once for the flux calculation.

### Finding the inductance

We use this result to find the inductance:

<!-- Equation labels as ordinary links -->
<div id="_auto17"></div>

$$
\begin{equation}
L = \frac{\Phi}{I} = \frac{h \mu_0 N^2}{2 \pi} \ln \frac{b}{a} \; .
\label{_auto17} \tag{17}
\end{equation}
$$

We notice that this is indeed a geometric relation. The flux only depends on geometric properties of the circuit and not on the current $I$.

### Approximation

What happens in the limit of a very thin toroid, where $b-a \ll a$? In this limit we can rewrite $b$ as $b-a+a$, getting:

<!-- Equation labels as ordinary links -->
<div id="_auto18"></div>

$$
\begin{equation}
\ln \frac{b}{a} = \ln \frac{b -a + a}{a} = \ln \left( 1 + \frac{b-a}{a}\right) \simeq \frac{b-a}{a}\; ,
\label{_auto18} \tag{18}
\end{equation}
$$

where we have used the Taylor expansion $\ln (1+x) \simeq x$, which is valid when $x \ll 1$. This gives us the result:

<!-- Equation labels as ordinary links -->
<div id="_auto19"></div>

$$
\begin{equation}
L \simeq \frac{h \mu_0 N^2}{2 \pi} \frac{d}{a} \; ,
\label{_auto19} \tag{19}
\end{equation}
$$

where $d = b-a$ is the thickness of the solenoid. We can rewrite this as $dh = S$ is the cross-sectional area of the toroid:

<!-- Equation labels as ordinary links -->
<div id="_auto20"></div>

$$
\begin{equation}
L \simeq \frac{S \mu_0 N^2}{2 \pi a } \; .
\label{_auto20} \tag{20}
\end{equation}
$$

## Example: Coaxial cable

A coaxial cable with an inner conductor with radius $a$ and an outer, thin conductor at radius $b$. Find the (self) inductance per unit length of the cable.

**Solution:**

Our plan is to first assume a current $I$ passes through the circuit, which is along the inner conductor and then back along the outer condutor. We find the magnetic field generated by the current, the flux of this current through the circuit and then use this to find the inductance.

Fig. [fig:inductance-coaxial-010](#fig:inductance-coaxial-010) illustrates the system. However, we need to decide on what we want to consider the circuit. When a current $I$ runs in the inner conductor, a corresponding current $I$ must run in the opposite direction in the outer conductor. We assume these two are connected by a radially directed wire as illustrated. We assume that the outer conductor is thin and that the current is limited to the surface. Since the system has cylinder symmetry, we will assume that the current is uniformly distributed on the surface.

### Finding the magnetic field

We find $\mathbf{B}$ using Ampere's law. The system has cylindrical symmetry. The field does not have any $z$-component, and cannot have a radial component, because this would lead to a finite flux of the magnetic field out of a cylindrical surface, and the flux must be zero. The field can therefore only have an azimuthal component. This component cannot depend on $z$ because we assume the system to be long, and it cannot depend on $\phi$, because the system is the same under any rotation around the cylinder axis. Therefore, $\mathbf{B} = B_{\phi}(r)\boldsymbol{\hat{\phi}} = B(r) \boldsymbol{\hat{\phi}}$.

We apply Ampere's law to the circuit illustrated in Fig. [fig:inductance-coaxial-010](#fig:inductance-coaxial-010)c:

<!-- Equation labels as ordinary links -->
<div id="_auto21"></div>

$$
\begin{equation}
\oint_C \mathbf{B} \cdot \mathrm{d} \mathbf{l} = B \, 2 \pi r = \mu_0 I \; ,
\label{_auto21} \tag{21}
\end{equation}
$$

when the circuit is within $a<r<b$. This gives

<!-- Equation labels as ordinary links -->
<div id="_auto22"></div>

$$
\begin{equation}
\mathbf{B} = \frac{\mu_0 I}{2 \pi r} \boldsymbol{\hat{\phi}} \; .
\label{_auto22} \tag{22}
\end{equation}
$$

### Finding the flux

We can then find the flux through the surface $S$, which is the flat surface spanned by the circuit in which the current $I$ is running as illustrated in the figure:

<!-- Equation labels as ordinary links -->
<div id="_auto23"></div>

$$
\begin{equation}
\Phi = \int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto23} \tag{23}
\end{equation}
$$

where a surface element goes from $r$ to $r + \mathrm{d} r$ with length $l$, and area $\mathrm{d} S= l \mathrm{d} r$:

<!-- Equation labels as ordinary links -->
<div id="_auto24"></div>

$$
\begin{equation}
\Phi = \int_S B \mathrm{d} S = \int_a^b B l \mathrm{d} r = \int_a^b \frac{\mu_0 I}{2 \pi r}l \mathrm{d} r 
\label{_auto24} \tag{24}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto25"></div>

$$
\begin{equation}  
= \frac{\mu_0 I}{2 \pi } \int_a^b \frac{\mathrm{d} r}{r} = \frac{\mu_0 I}{2 \pi } l \ln \frac{b}{a} \; .
\label{_auto25} \tag{25}
\end{equation}
$$

### Finding the inductance

Finally, we find the inductance $L$ from

<!-- Equation labels as ordinary links -->
<div id="_auto26"></div>

$$
\begin{equation}
L = \frac{\Phi}{I} = \frac{\mu_0 I l}{2 \pi I} \ln \frac{b}{a} =
\frac{\mu_0 l}{2 \pi} \ln \frac{b}{a}
\label{_auto26} \tag{26}
\end{equation}
$$

and

<!-- Equation labels as ordinary links -->
<div id="_auto27"></div>

$$
\begin{equation}
\frac{L}{l} = \frac{\mu_0}{2 \pi} \ln \frac{b}{a}
\label{_auto27} \tag{27}
\end{equation}
$$

# Mutual inductance

We have so far defined inductance by how a circuit induces a flux on itself. However, a circuit can also incude a flux in another circuit. We therefore generalize the term inductance to *mutual inductance*. Fig. [fig:inductance-mutual-010](#fig:inductance-mutual-010) illustrates two circuits $C_1$ and $_C2$. The current $I_1$ in generates a magnetic field $\mathbf{B}_1$, and the flux of the magnetic field $\mathbf{B}_1$ thorugh $C_2$ is called

<!-- Equation labels as ordinary links -->
<div id="_auto28"></div>

$$
\begin{equation}
\Phi_{1,2} = \int_{S_2} \mathbf{B}_1\cdot \mathrm{d} \mathbf{S} \; .
\label{_auto28} \tag{28}
\end{equation}
$$

The first index refers to the source of the flux. Here it is circuit $C_1$. The second index refers to the circuit through which we measure the flux. Here it is a surface $S_2$ enclosed by $C_2$. This is the flux from circuit 1 in circuit 2. We define the mutual inductance $L_{1,2}$ as

<!-- Equation labels as ordinary links -->
<div id="_auto29"></div>

$$
\begin{equation}
L_{1,2} = \frac{\Phi_{1,2}}{I_1} \; ,
\label{_auto29} \tag{29}
\end{equation}
$$

using the same interpretation of the indexes as for the flux. We recognize the inductances $L_{1,1}$ and $L_{2,2}$ as the self inductances of circuits $C_1$ and $C_2$ respectively. There are in general four mutual inductances $L_{1,1}$, $L_{1,2}$, $L_{2,1}$, and $L_{2,2}$.

## Definition of mutual inductance

**Mutual inductance.**

We define the flux in a circuit $C_j$ from a current $I_i$ in circuit $C_i$ to be

<!-- Equation labels as ordinary links -->
<div id="_auto30"></div>

$$
\begin{equation}
\Phi_{i,j} = \int_{S_j} \mathbf{B} \cdot \mathrm{d} \mathbf{S} \; 
\label{_auto30} \tag{30}
\end{equation}
$$

where $S_j$ is a surface enclosed by $C_j$. We define the **mutual inductanc** $L_{i,j}$ as

<!-- Equation labels as ordinary links -->
<div id="_auto31"></div>

$$
\begin{equation}
L_{i,j} = \frac{\Phi_{i,j}}{I_i} \; .
\label{_auto31} \tag{31}
\end{equation}
$$

## Symmetry of mutual inductances

A very practical result, which we will often use, relates the two mutual inductances of two circuits $C_1$ and $C_2$:

<!-- Equation labels as ordinary links -->
<div id="_auto32"></div>

$$
\begin{equation}
L_{1,2} = L_{2,1}
\label{_auto32} \tag{32}
\end{equation}
$$

This nice symmetry provides us with a very powerful tool: If we want to calculate the mutual inductance, we can choose to calculate the magnetic field from the circuit where this is the simplest, and use this to find the flux through the other circuit. We will demonstrate how we use this principle below.

### Proof

There is a very nice proof of this principle, which depends on the vector potential $\mathbf{A}$. You may skip this proof without loss of continuity.

The flux through loop $C_2$ from the magnetic field set up by circuit $C_1$ is:

<!-- Equation labels as ordinary links -->
<div id="_auto33"></div>

$$
\begin{equation}
\Phi_{1,2} = \int_{S} \mathbf{B}_1 \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto33} \tag{33}
\end{equation}
$$

where we insert the vector potential $\mathbf{A}_1$ for $\mathbf{B}_1$:

<!-- Equation labels as ordinary links -->
<div id="_auto34"></div>

$$
\begin{equation}
\Phi_{1,2} = \int_S \left( \nabla \times \mathbf{A}_1 \right) \cdot \mathrm{d} \mathbf{S} \; ,
\label{_auto34} \tag{34}
\end{equation}
$$

and we can then apply Stoke's theorem to get:

<!-- Equation labels as ordinary links -->
<div id="_auto35"></div>

$$
\begin{equation}
\Phi_{1,2} = \int_S \left( \nabla \times \mathbf{A}_1 \right) \cdot \mathrm{d} \mathbf{S}  = \int_{C_2} \mathbf{A}_1 \mathrm{d} \mathbf{l}_2 \; .
\label{_auto35} \tag{35}
\end{equation}
$$

The vector potential $\mathbf{A}_1$ is found from:

<!-- Equation labels as ordinary links -->
<div id="_auto36"></div>

$$
\begin{equation}
\mathbf{A}_1 = \frac{\mu_0 I_1}{4 \pi} \oint_{C_1} \frac{\mathrm{d} \mathbf{l}_1}{R}
\label{_auto36} \tag{36}
\end{equation}
$$

We insert this into the expression for the flux, and get:

<!-- Equation labels as ordinary links -->
<div id="_auto37"></div>

$$
\begin{equation}
\Phi_{1,2} = \frac{\mu_0 I_1}{4 \pi}\oint_{C_2} \left( \oint_{C_1} \frac{\mathrm{d} \mathbf{l_1}}{R} \right) \cdot \mathrm{d} \mathbf{l}_2 \; .
\label{_auto37} \tag{37}
\end{equation}
$$

We therefore find that the mutual inductance is

<!-- Equation labels as ordinary links -->
<div id="_auto38"></div>

$$
\begin{equation}
L_{1,2} = \frac{\mu_0}{4 \pi} \oint_{C_1} \oint_{C_2} \frac{\mathrm{d} \mathbf{l}_1 \cdot \mathrm{d} \mathbf{l}_2}{R}
\label{_auto38} \tag{38}
\end{equation}
$$

This is called the *Neumann formula*. The order in which we perform this double line integral is arbitrary. This proves that $L_{1,2} = L_{2,1}$. And it also demonstates that the mutual inductance is a purely geometry property: The line integrals depend on the shapes of the curves, but not on the currents in the circuits.

Fig. [fig:inductance-mutualtoroid-010](#fig:inductance-mutualtoroid-010) shows a toroid with a rectangular cross section with a wire wrapped around the whole toiroid ($C_1$) and a single current loop $C_2$. What is the mutual inductance $L_{2,1}$ of this system?

**Solution:**

In order to find $L_{2,1}$ we would need to find the magnetic field $\mathbf{B}_2$ generated by the current $I_2$ in circuit $C_2$. This is not simple. However, we can use that $L_{2,1} = L_{1,2}$. Finding $L_{1,2}$ is simpler, because it is simpler to find the magnetic field from the toroid circuit $C_1$.

We assume that the current $I_1$ is in the toroid. What is the magnetic field inside the toroid. We have previously found that

<!-- Equation labels as ordinary links -->
<div id="_auto39"></div>

$$
\begin{equation}
\mathbf{B}_1 = \frac{\mu N_1 I_1}{2 \pi r} \boldsymbol{\hat{\phi}} \; ,
\label{_auto39} \tag{39}
\end{equation}
$$

where $N_1$ is the number of times the wire is wrapped around the toroid and $\mu$ is the permeability of the magnetic material inside the toroid.

What is the flux of $\mathbf{B}_1$ through $S_2$? We notice that the magnetic field from the toroid is only non-zero inside the toroid and zero outside. It is therefore only the intersection of the surface $S_2$ and the toroid that contributes to the integral. We choose a surface which is normal to the toroid, so that the intersection is a rectangular intersection from $r = a$ to $r=b$ and with height $h$. (We can choose such a surface, because we are free to choose any surface $S_2$ which is enclosed by $C_2$.)

The of $\mathbf{B}_1$ through $C_2$ is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto40"></div>

$$
\begin{equation}
\Phi_{1,2} = \int_{C_2}\mathbf{B}_1 \cdot \mathrm{d} \mathbf{S} = \int_a^b B h \mathrm{d} r = \int_a^b \frac{h \mu N_1}{2 \pi r} \mathrm{d} r = \frac{\mu h N_1}{2 \pi} \ln \frac{b}{a} .
\label{_auto40} \tag{40}
\end{equation}
$$

If instead circuit $C_2$ consisted of $N_2$ windings, we would get $N_2$ times the flux from one winding:

<!-- Equation labels as ordinary links -->
<div id="_auto41"></div>

$$
\begin{equation}
\Phi_{1,2} =  \frac{\mu h N_1 N_2}{2 \pi} \ln \frac{b}{a} \; .
\label{_auto41} \tag{41}
\end{equation}
$$

# Applications of mutual inductance

If a time-varying current $I_1$ is sent through circuit $C_1$ in Fig. [fig:inductance-mutual-010](#fig:inductance-mutual-010), an emf $e_{1,2}$ will be induced in circuit $C_2$. This emf will be:

<!-- Equation labels as ordinary links -->
<div id="_auto42"></div>

$$
\begin{equation}
e_{1,2} = -\frac{\mathrm{d} \Phi_{1,2}}{\mathrm{d} t} = - \frac{\mathrm{d} \left( L_{1,2}I_1 \right)}{\mathrm{d} t} \; .
\label{_auto42} \tag{42}
\end{equation}
$$

If the loops do not change with time, then $\mathrm{d} L_{1,2}/ \mathrm{d} t = 0$, and

<!-- Equation labels as ordinary links -->
<div id="_auto43"></div>

$$
\begin{equation}
e_{1,2} = -L_{1,2} \frac{\mathrm{d} I_1}{\mathrm{d} t} \; .
\label{_auto43} \tag{43}
\end{equation}
$$

This opens for a wireless transmission of voltage, current and energy between circuits. We will here address a few examples of this interaction such as transformers, but this type of interaction is the basis of many technologies.

**Solution:**

We follow our standard method for calculating the mutual inductance. We assume current $I_1$ is running in $C_1$ and find the flux through $C_2$. We realize that the magnetic field from $C_1$ is due to the superposition of two fields, one field from each of the two wires in $C_1$. We find the fluxes from each magnetic field and sum them to find the total flux from $C_1$ through a surface spanned by $C_2$.

### Finding the mutual conductance

We start by finding the field from the current coming out of the plane in the top left conductor in  Fig. [fig:inductance-mutual-010](#fig:inductance-mutual-010). This is a current-carrying, long linear wire. We find the magnetic field by applying Ampere's law on a circular path of radius $r$ around the wire:

<!-- Equation labels as ordinary links -->
<div id="_auto44"></div>

$$
\begin{equation}
\oint_C \mathbf{B}_1 \cdot \mathrm{d} \mathbf{l} = 2 \pi r B_1 = \mu_0 I_1 \quad \Rightarrow \quad B_1 = \frac{\mu_0 I_1}{2 \pi r}\; ,
\label{_auto44} \tag{44}
\end{equation}
$$

where $r$ is the distance to the wire.

Second, we want to find the flux of $B_1$ through a surface enclosed by the two wires in $C_2$. We are free to choose any surface that is convenient for the integration. We therefore choose to divide the surface into two parts, $S_1$ and $S_2$ as shown in the figure. The surface $S_1$ is a cylindrical surface around the wire. Because the magnetic field is azimuthal to the wire, it curls around the wire, the magnetic field is tangential to this surface and the flux is zero. Surface $S_2$ is directed in the radial direction. The magnetic field is therefore normal to $S_2$ and the flux integral is simple to find and very similar to the integral we found for the toroid system:

<!-- Equation labels as ordinary links -->
<div id="_auto45"></div>

$$
\begin{equation}
\Phi_A = \underbrace{\int_{S_1} \mathbf{B}_1 \cdot \mathrm{d} \mathbf{S}}_{=0} + \int_{S_2} \mathbf{B}_1 \cdot \mathrm{d} \mathbf{S} = \int_{r=a}^{r=d} \frac{\mu_0 I_1}{2 \pi r}l\mathrm{d} r = \frac{\mu_0 I_1 l}{2 \pi}\ln \frac{d}{a} \; ,
\label{_auto45} \tag{45}
\end{equation}
$$

where $l$ is the length into the plane of the part of the wire we analyze. We find a similar result for the flux from the other wire, but the direction of the current is in the opposite direction:

<!-- Equation labels as ordinary links -->
<div id="_auto46"></div>

$$
\begin{equation}
\Phi_B = - \frac{\mu_0 I_1 l}{2 \pi} \frac{b}{c} \; .
\label{_auto46} \tag{46}
\end{equation}
$$

The total flux is the sum:

<!-- Equation labels as ordinary links -->
<div id="_auto47"></div>

$$
\begin{equation}
\Phi = \frac{\mu_0 I_1 l}{2 \pi}\ln \frac{d}{a} - \frac{\mu_0 I_1 l}{2 \pi} \frac{b}{c} =
\frac{\mu_0 I_1 l}{2 \pi}\ln \frac{d c}{a b} \; .
\label{_auto47} \tag{47}
\end{equation}
$$

The mutual inductance per unit length is therefore

<!-- Equation labels as ordinary links -->
<div id="_auto48"></div>

$$
\begin{equation}
L_{1,2} = \frac{\Phi}{l I_1} = \frac{\mu_0}{2 \pi}\ln \frac{d c}{a b}
\label{_auto48} \tag{48}
\end{equation}
$$

### Understanding cross-talk

We can use this formula to understand the inductive coupling between two wires, such as between two wires in an cable or a an electric devices. The current in one wire pair, will induce a emf in the other wire pair, which will impact the signales carried in the wires. Your internet cable consists of pairs of wires carrying high-frequency currents, and inductive cross-talk should be reduce to reduce interaction between signals in different wires.

## Example: Voltage and current transformation

Mutual inductance allows the coupling between two circuits and can be used to transform voltages or currents. Here, we will study a typical system used for transformation and how it can be used to transform voltage and current.

Fig. [fig:inductance-transformer-010](#fig:inductance-transformer-010) illustrates a typical geometry for a transformer. The system consists of a magnetic circuit formed by a thin toroid with a ferromagnetic core and two windings, that is, two current loops, $C_1$ and $C_2$ with $N_1$ and $N_2$ windings respectively.

(b) Circuit diagram model of the same system. (c) Simplified circuit diagram for one circuit. <div id="fig:inductance-transformer-010"></div> -->
<!-- begin figure -->
<div id="fig:inductance-transformer-010"></div>

<p>(a) Illustration of a transformer consisting of two circuits wrapped around a ferromagnetic toroid. (b) Circuit diagram model of the same system. (c) Simplified circuit diagram for one circuit.</p>
<img src="fig-inductance/inductance-transformer-010.png" width=600>

<!-- end figure -->


The emf induced in circuit 1 from $I_1$ is called $e_{1,1}$ and the emf induced in circuit $C_2$ from current $I_1$ is $e_{1,2}$. We follow the same indexing convention as before: The first index refers to the circuit that generates the magnetic field and the second index refers to the circuit where the flux is generated or the emf induced. We know that

<!-- Equation labels as ordinary links -->
<div id="_auto49"></div>

$$
\begin{equation}
\frac{e_{1,2}}{e_{1,1}} = \frac{ -\frac{\mathrm{d} \Phi_{1,2}}{\mathrm{d} t}} { - \frac{\mathrm{d} \Phi_{1,1}}{\mathrm{d} t}} = \frac{L_{1,1}\frac{\mathrm{d} I_1}{\mathrm{d} t}}{L_{1,1}\frac{\mathrm{d} I_1}{\mathrm{d} t}} = \frac{L_{1,2}}{L_{1,1}} \; .
\label{_auto49} \tag{49}
\end{equation}
$$

This is a general result which is not based on any assumptions for arbitrary circuits. We can use this result to address the ratio of the induced emfs based on calculated or measured values for the mutual and self inductances.

### Ideal transformer

We start by addressing an ideal transformer where $\mu = \infty$ so that all the flux remains in the ferromagnet. We want to find the ratio between the voltages $V_1$ on the signal generator and the voltage $V_2$ across the resistor in circuit $C_2$.

In these systems it is often tricky to find the correct signs for the voltages. Let us look at the simplified circuit in Fig. [fig:inductance-transformer-010](#fig:inductance-transformer-010)c. For this circuit we know that $\sum e = RI$, which is

<!-- Equation labels as ordinary links -->
<div id="_auto50"></div>

$$
\begin{equation}
V_b + \left( - \frac{\mathrm{d} \Phi}{\mathrm{d} t} \right) = RI \; .
\label{_auto50} \tag{50}
\end{equation}
$$

In the case when $R = 0$, which is the case on the left in the figure (circuit $C_1$), we get that $V_b = V_1 = \mathrm{d} \Phi/ \mathrm{d} t$. In the case when $V_b = 0$, which is the case on the right in the figure (circuit $C_2$), we get that $RI = - \mathrm{d} \Phi / \mathrm{d} t$.

### Voltage transformation

This ratio between $V_2$ and $V_1$ is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto51"></div>

$$
\begin{equation}
\frac{V_2}{V_1} = \frac{R I_2}{V_1} = \frac{ - \frac{\mathrm{d} \Phi_2}{\mathrm{d} t}}{+ \frac{\mathrm{d} \Phi_1}{\mathrm{d} t}}
\label{_auto51} \tag{51}
\end{equation}
$$

What are the fluxes $\Phi_1$ and $\Phi_2$. These depend on the cross-sectional flux, $\Phi_{CS}$, through a cross-section of the ferromagnetic materials.  We therefore have that $|\Phi_2| = N_2 \Phi_{CS}$ and $|\Phi_1| = N_1 \Phi_{CS}$, because each of the circuits wraps $N_1$ and $N_2$ times around the ferromagnet respectively. Let us also check that the signs of these fluxes. The cross-sectional flux is positive as indiciated in Fig. [fig:inductance-transformer-010](#fig:inductance-transformer-010)a. The flux $\Phi_1$ should be positive, so that $\Phi_1 = N_1 \Phi_{CS}$. However, we see that $\Phi_2 <0$, so that $\Phi_2 = - N_2 \Phi_{CS}$.

We insert this in the ratio and find

<!-- Equation labels as ordinary links -->
<div id="_auto52"></div>

$$
\begin{equation}
\frac{V_2}{V_1} =\frac{ - \frac{\Phi_2}{\mathrm{d} t}}{+ \frac{\mathrm{d} \Phi_1}{\mathrm{d} t}} =
\frac{ \frac{\mathrm{d} N_2 \Phi_{CS}}{\mathrm{d} t}}{\frac{\mathrm{d} N_1 \Phi_{CS}}{\mathrm{d} t}} = \frac{N_2}{N_1} \; .
\label{_auto52} \tag{52}
\end{equation}
$$

This system therefore acts as a transformer and can change one voltage into another voltage, while transmitting the signal.

$$
\begin{equation}
\int_C \mathbf{H} \cdot \mathrm{d} \mathbf{l} = N_1 I_1 - N_2 I_2 \; .
\label{_auto53} \tag{53}
\end{equation}

In the case when $\mu \rightarrow \infty$ we must therefore have that $H \rightarrow 0$, because the energy density cannot be zero. This is of course only an approxiation, because the real $\mu$ is not infinity, but still a large number. Notice that this does not imply that $\mathbf{B}$ is zero. We then get that:

# Energy in magnetic fields

For capacitors, we developed a theory for the energy needed to charge a capacitor to a given charge and used this to define the energy density of the electric field. Here, we will follow a similar approach to define the energy density of the magnetic field.

## Energy in a single circuit

What is the energy needed to get a current $I$ flowing in a circuit such as the circuit in Fig. [fig:inductance-energy-010](#fig:inductance-energy-010)? You may think it is zero, because you can just turn a switch, and then the current is running. However, if you start with zero current, you need to gradually change the current until you reach a current $I$. As you do this, you will experience an emf that will counteract the change in current. For the circuit in the figure, Kirchoff's voltage law gives:

<!-- Equation labels as ordinary links -->
<div id="_auto58"></div>

$$
\begin{equation}
V_b = R I - e \; ,
\label{_auto58} \tag{58}
\end{equation}
$$

where $V_b$ is the emf of the battery and $e$ is the emf from the self-inductance of the circuit. The power $P$ consumed in this circuit is

<!-- Equation labels as ordinary links -->
<div id="_auto59"></div>

$$
\begin{equation}
P = V_b I = R I^2- e I \; ,
\label{_auto59} \tag{59}
\end{equation}
$$

and the work done in a short time interval $\mathrm{d} t$ is

<!-- Equation labels as ordinary links -->
<div id="_auto60"></div>

$$
\begin{equation}
\mathrm{d} W = P \mathrm{d} t = V_b I \mathrm{d} t = R I^2 \mathrm{d} t  - e I \mathrm{d} t \; .
\label{_auto60} \tag{60}
\end{equation}
$$

The term $V_b I \mathrm{d} t$ represents the energy supplied by the battery. Some of this energy is dissipated in the resistor. This is the term $R I^2 \mathrm{d} t$. The remaining energy is used as work against the emf due to the self inductance of the circuit. This is the energy that is stored in the magnetic field in the circuit. The work done by external sources --- the battery --- to the magnetic field of the circuit is

<!-- Equation labels as ordinary links -->
<div id="_auto61"></div>

$$
\begin{equation}
\mathrm{d} W_m = - e I \mathrm{d} t = - \left(- \frac{\mathrm{d} \Phi}{\mathrm{d} t} \right) I \mathrm{d} t 
\label{_auto61} \tag{61}
\end{equation}
$$

We insert $\Phi = LI$, so that $\mathrm{d} \Phi / \mathrm{d} t = L \mathrm{d} I / \mathrm{d} t$ (when $L$ is not changing with time), getting

<!-- Equation labels as ordinary links -->
<div id="_auto62"></div>

$$
\begin{equation}
\mathrm{d} W_m = L I \frac{\mathrm{d} I}{\mathrm{d} t} = L I \mathrm{d} I \; .
\label{_auto62} \tag{62}
\end{equation}
$$

This is the work done by the external source to change the current by an amount $\mathrm{d} I$. If we bring the system from zero current, where the magnetic energy is zero, up to a current $I$, the work done on the magnetic field is:

<!-- Equation labels as ordinary links -->
<div id="_auto63"></div>

$$
\begin{equation}
W_m = \int_0^I L I \mathrm{d} I = \frac{1}{2} L I^2 \; .
\label{_auto63} \tag{63}
\end{equation}
$$

We can rewrite this using the flux $\Phi = LI$, getting:

<!-- Equation labels as ordinary links -->
<div id="_auto64"></div>

$$
\begin{equation}
W_m = \frac{1}{2}LI^2 = \frac{1}{2}\Phi I = \frac{\Phi^2}{2 L} \; .
\label{_auto64} \tag{64}
\end{equation}
$$

This is similar to the expressions we found for a capacitor, where $W_e = QV/2 = CV^2/2$.

<!-- begin figure -->
<div id="fig:inductance-energy-010"></div>

<img src="fig-inductance/inductance-energy-010.png" width=600>

<!-- end figure -->


**Energy stored in an inductor.**

The energy $W_m$ stored in the magnetic field of an inductor with inductance $L$ and a current $I$ when it does not interact magnetically with any other circuit is:

<!-- Equation labels as ordinary links -->
<div id="_auto65"></div>

$$
\begin{equation}
W_m = \frac{1}{2} L I^2 = \frac{1}{2} I \Phi \; .
\label{_auto65} \tag{65}
\end{equation}
$$

## Energy in coupled circuits

This result can be generalized to a system with $N$ magnetically coupled circuit loops. The total energy is

<!-- Equation labels as ordinary links -->
<div id="eq:inductance-energy-010"></div>

$$
\begin{equation}
W_m = \frac{1}{2} \sum_{i=1}^N I_i \Phi_i \; ,
\label{eq:inductance-energy-010} \tag{66}
\end{equation}
$$

where the flux $\Phi_i$ through circuit $i$ is

<!-- Equation labels as ordinary links -->
<div id="_auto66"></div>

$$
\begin{equation}
\Phi_i = \sum_{j=1}^N \Phi_{j,i} \; , 
\label{_auto66} \tag{67}
\end{equation}
$$

is the sum of the fluxes from a circuit $j = 1,\ldots, N$ on a circuit $i$, where $\Phi_{j,i} = L_{j,i} I_i$. The magnetic energy is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto67"></div>

$$
\begin{equation}
W_m = \frac{1}{2}\sum_{i=1}^N \sum_{j=1}^N L_{j,i} I_i I_j \; .
\label{_auto67} \tag{68}
\end{equation}
$$

The magnetic energy of two coupled loops can therefore be written as:

<!-- Equation labels as ordinary links -->
<div id="_auto68"></div>

$$
\begin{equation}
W_m = \frac{1}{2} \left(  \underbrace{L_{1,1} I_1 I_1 + L_{2,1} I_2 I_1}_{\Phi_1} +  \underbrace{L_{1,2} I_1 I_2 + L_{2,2} I_2 I_ 2}_{\Phi_2} \right)
\label{_auto68} \tag{69}
\end{equation}
$$

Notice that $L_{1,1}$ and $L_{2,2}$ are positive, but that $L_{1,2}$ can be both positive or negative. The total energy of a coupled system can therefore both be larger or smaller than the sum of the energies of the individual systems.

**Energy stored in an set of coupled inductors.**

The energy $W_m$ stored in the magnetic field of a set of $N$ magnetically coupled inductors with mutual inductances $L_{i,j}$ and a current $I_i$ isother circuit is:

<!-- Equation labels as ordinary links -->
<div id="_auto69"></div>

$$
\begin{equation}
W_m = \frac{1}{2} \sum_i \sum_j L_{i,j} I_i I_j \; .
\label{_auto69} \tag{70}
\end{equation}
$$

### Proof for a set of coupled circuits

For completeness, we provide a proof of ([66](#eq:inductance-energy-010)). If we have $N$ coupled circuits, then Kirchoff's voltage law for circuit $j$ is

<!-- Equation labels as ordinary links -->
<div id="_auto70"></div>

$$
\begin{equation}
V_{b,j} + \left( - \frac{\mathrm{d} \Phi_j}{\mathrm{d} t} \right) = R_j I_j \; ,
\label{_auto70} \tag{71}
\end{equation}
$$

where $V_{b,j}$ is the emf of the voltage source in circuit $j$. The power dissipated in circuit $j$ is then:

<!-- Equation labels as ordinary links -->
<div id="_auto71"></div>

$$
\begin{equation}
P_j = V_{b,j} I_j = \left( R I_j + \frac{\mathrm{d} \Phi_j}{\mathrm{d} t} \right) I_k = R I_j^2 + \frac{\mathrm{d} \Phi_j}{\mathrm{d} t}I_j \; ,
\label{_auto71} \tag{72}
\end{equation}
$$

where we recognize the first term as the power dissipation in the resistor and the second term is related to the energy stored in the magnetic field. The work delivered by voltage source $j$ in a time interval $\mathrm{d} t$ is then:

<!-- Equation labels as ordinary links -->
<div id="_auto72"></div>

$$
\begin{equation}
\mathrm{d} W_j = R_j I_j^2 + I_j \mathrm{d} \Phi_j \; .
\label{_auto72} \tag{73}
\end{equation}
$$

The total magnetic work done by all the voltage sources in a time interval $\mathrm{d} t$ is

<!-- Equation labels as ordinary links -->
<div id="_auto73"></div>

$$
\begin{equation}
\mathrm{d} W_m = \sum_j I_j \mathrm{d} \Phi_j \; ,
\label{_auto73} \tag{74}
\end{equation}
$$

where

<!-- Equation labels as ordinary links -->
<div id="_auto74"></div>

$$
\begin{equation}
\Phi_j = \sum_i \Phi_{i,j} = \sum_i L_{i,j} I_i \; .
\label{_auto74} \tag{75}
\end{equation}
$$

When $L_{i,j}$ does not change with time we have that $\mathrm{d} (L_{i,j} I_i ) = L_{i,j} \mathrm{d} I_i$:

<!-- Equation labels as ordinary links -->
<div id="eq:inductance-energy-020"></div>

$$
\begin{equation}
\mathrm{d} W_m = \sum_i \sum_j L_{i,j} I_j \mathrm{d} I_i \; ,
\label{eq:inductance-energy-020} \tag{76}
\end{equation}
$$

We now use the trick that $\mathrm{d} (I_i I_j) = I_i \mathrm{d} I_j + I_j \mathrm{d} I_i$ and that $L_{i,j} = L_{j,i}$ so that

<!-- Equation labels as ordinary links -->
<div id="_auto75"></div>

$$
\begin{equation}
\mathrm{d} \left( \sum_i \sum_j L_{i,j} I_i I_j \right) = \sum_i \sum_j L_{i,j} \left( I_j \mathrm{d} I_i + \mathrm{d} I_i I_j  \right) 
\label{_auto75} \tag{77}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto76"></div>

$$
\begin{equation}  
= \sum_i \sum_j L_{i,j} I_j \mathrm{d} I_i + \sum_i \sum_j L_{j,i} I_i \mathrm{d} I_j 
\label{_auto76} \tag{78}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto77"></div>

$$
\begin{equation}  
= 2 \sum_i \sum_j L_{i,j} I_j \mathrm{d} I_i \; ,
\label{_auto77} \tag{79}
\end{equation}
$$

We therefore see that with

<!-- Equation labels as ordinary links -->
<div id="_auto78"></div>

$$
\begin{equation}
W_m = \frac{1}{2}\sum_i \sum_j L_{i,j} I_i I_j \; .
\label{_auto78} \tag{80}
\end{equation}
$$

we get the relation in ([76](#eq:inductance-energy-020)), which proves that this is the magnetic energy of the system.

## Energy density
<div id="sec:induction-energydensity"></div>

What is the energy in a thin toroid with $N$ windings expressed in terms of the magnetic field $B$, $H$? Fig. [fig:inductance-energytoroid-010](#fig:inductance-energytoroid-010) illustrates the system. The energy stored in this system is

<!-- Equation labels as ordinary links -->
<div id="_auto79"></div>

$$
\begin{equation}
W_m = \frac{1}{2}L I^2 = \frac{1}{2} I \Phi \; .
\label{_auto79} \tag{81}
\end{equation}
$$

We can relate the current $I$ to the magnetic field in the toroid by applying Ampere's law on a circuit $C$ around the center of the toroid, where the magnetic $\mathbf{H}$-vector is in the azimuthal direction due to symmetry, $\mathbf{H} = H(r) \boldsymbol{\hat{\phi}}$.  Ampere's law gives

<!-- Equation labels as ordinary links -->
<div id="_auto80"></div>

$$
\begin{equation}
\oint_C \mathbf{H} \cdot \mathrm{d} \mathbf{l} = 2 \pi r H = N I \quad \Rightarrow \quad I = \frac{2 \pi r H}{N} \; .
\label{_auto80} \tag{82}
\end{equation}
$$

The flux $\Phi$ of this magnetic field through the circuit $C$ is $N$ times the flux through a cross section with area $S$. If the toroid is thin compared to its radius, we can assume that the magnetic field is approximately uniform across the cross section and the flux through a cross-section $S$ is:

<!-- Equation labels as ordinary links -->
<div id="_auto81"></div>

$$
\begin{equation}
\Phi_S = \int_S \mathbf{B} \cdot \mathrm{d} \mathbf{S} = B S \; 
\label{_auto81} \tag{83}
\end{equation}
$$

and the total flux through $N$ windings is $\Phi = N \Phi_S = N B S$. The magnetic energy stored in this system is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto82"></div>

$$
\begin{equation}
W = \frac{1}{2} I \Phi = \frac{1}{2} \left( \frac{2 \pi r H}{N} \right) B S N = \frac{1}{2} 2 \pi r S B H = \frac{1}{2} B H \, 2 \pi r S \; .
\label{_auto82} \tag{84}
\end{equation}
$$

Here, we recognize the term $BH/2$ as the energy per unit volume and $2 \pi r S$ as the volume of the toroid. We can therefore introduce the energy density $u$

<!-- Equation labels as ordinary links -->
<div id="_auto83"></div>

$$
\begin{equation}
u = \frac{W}{V} = \frac{1}{2}B H \; ,
\label{_auto83} \tag{85}
\end{equation}
$$

This result is general, but we do not provide a general proof of this result here. (This proof typically requires the use of the vector potential for the magnetic field).

**Energy density of a magnetic field.**

The **energy density** $u(\mathbf{r})$ of a magnetic field described by $\mathbf{B}$ and $\mathbf{H}$ is

<!-- Equation labels as ordinary links -->
<div id="_auto84"></div>

$$
\begin{equation}
u = \frac{1}{2} \mathbf{B} \cdot \mathbf{H} \; .
\label{_auto84} \tag{86}
\end{equation}
$$

## Magnetic energy in a non-linear medium

What happens in a non-linear magnetic material? This is an important practical question, because we often use non-linear ferromagnetic cores in magnetic circuits and in various coils (toroids) in electric circuits.

The work performed by an external field (e.g. a voltage source or a battery) to change the magnetic flux in a magnetic circuit is:

<!-- Equation labels as ordinary links -->
<div id="_auto85"></div>

$$
\begin{equation}
\mathrm{d} W_m = I \mathrm{d} \Phi \; .
\label{_auto85} \tag{87}
\end{equation}
$$

For a thin toroid of radius $r$ we know from Ampere's law that the magnetic field $H$ inside the toroid is

<!-- Equation labels as ordinary links -->
<div id="_auto86"></div>

$$
\begin{equation}
H = \frac{N I }{2 \pi r} \; ,
\label{_auto86} \tag{88}
\end{equation}
$$

and that the total flux through the circuit wrapping around the toroid is $\Phi = N B S$. The contributions to the work is then:

<!-- Equation labels as ordinary links -->
<div id="_auto87"></div>

$$
\begin{equation}
\mathrm{d} W_m = \underbrace{\frac{2 \pi r H}{N}}_{I} \, \underbrace{NS \mathrm{d} B}_{\mathrm{d} \Phi} = \underbrace{2 \pi r S}_{\text{volume}} \, H \mathrm{d} B \; .
\label{_auto87} \tag{89}
\end{equation}
$$

The change in energy density is the work done per unit volume, which is

<!-- Equation labels as ordinary links -->
<div id="_auto88"></div>

$$
\begin{equation}
\mathrm{d} u = H \mathrm{d} B \; .
\label{_auto88} \tag{90}
\end{equation}
$$

So far, this is a general result. However, a non-linear medium has hysteresis. In Fig. [fig:inductance-nonlinear-010](#fig:inductance-nonlinear-010) we illustrate how the magnetic fields $B$ and $H$ are related during a complete cycle for an alternating current. The work per unit volume done on the magnetic system if we change the field from $B$ to $B + \mathrm{d} B$ is the area of a rectangle with height $\mathrm{d} B$ and length $H$.  In part $1$ of the path the changes in magnetic field are positive, but in part 2 the changes are negative, so that the work per unit volume is also negative. We therefore subtract the gray area. The work done on the magnetic field over a hole period is the area enclosed by the hysteresis loop. This work is energy which is dissipated as the orientations of the magnetic domains in the material and is transformed to thermal energy (heat) in the material.

<!-- dom:FIGURE:[fig-inductance/inductance-nonlinear-010.png, width=600 frac=0.55] Illustration of a hysteresis loop for the magnetic field in a circuit. (a) The integral of $H \mathrm{d} B$ along curve $1$ is positive. (b) The integral of $H \mathrm{d} B$ along curve $2$ is negative. (c) The net integral is the shaded area enclosed by the hysteresis curve. <div id="fig:inductance-nonlinear-010"></div> -->
<!-- begin figure -->
<div id="fig:inductance-nonlinear-010"></div>

<p>Illustration of a hysteresis loop for the magnetic field in a circuit. (a) The integral of $H \mathrm{d} B$ along curve $1$ is positive. (b) The integral of $H \mathrm{d} B$ along curve $2$ is negative. (c) The net integral is the shaded area enclosed by the hysteresis curve.</p>
<img src="fig-inductance/inductance-nonlinear-010.png" width=600>

The total area is the dissipated energy per cycle. If the cycle has a period $T$ and a frequency $f = 1/T$, the dissipated energy per unit time (and unit volume), the dissipated power per unit volume, is

<!-- Equation labels as ordinary links -->
<div id="_auto89"></div>

$$
\begin{equation}
p = \frac{P}{v} = \frac{\int H \mathrm{d} B}{v T} = A \, f \; ,
\label{_auto89} \tag{91}
\end{equation}
$$

where $A$ is the area of the hysteresis curve and $f$ is the frequency. In order to reduce the power dissipation, we should therefore try to design a system with as small an area of the hysteresis loop as possible (and have low frequencies).

## Example: Induction stove

In an induction stove, a magnetic field varying with a frequency $f$ enters the cooking pan. If the pan is made from a ferromagnetic material, with a given hysteresis curve, there will be a power dissipation which produces heat in the ferromagnetic material. This is main heating effect. In addition, there are eddy currents induced by the time-varying magnetic field, but the dissipation from these are usually small compared to the hysteresis effect. This is why aluminum is not an effective material for induction stove cooking pans --- aluminum is not a ferromagnet.

## Example: Coaxial cable

Find the energy per length stored in a coaxial cable carrying a current $I$. The inner radius of the cable is $a$ and the outer radius is $b$.

**Solution:**

We can find the energy either from $W_m = L I^2 /2$, that is, by finding $L$, or from $W_m = \int_V \mathbf{B} \cdot \mathbf{H} \mathrm{d} v$. We demonstrate the second approach.

First, we find the magnetic field using Ampere's law as we have done many times before. Due to the cylindrical symmetry and that the divergence of the magnetic field is zero, $\mathbf{H}$ and $\mathbf{B}$ only has an azimuthal component, $\mathbf{H} = H(r) \boldsymbol{\hat{\phi}}$. Ampere's law states

<!-- Equation labels as ordinary links -->
<div id="_auto90"></div>

$$
\begin{equation}
\int_C \mathbf{H} \cdot \mathrm{d} \mathbf{l} = 2 \pi r H = I \quad (a<r<b) \; ,
\label{_auto90} \tag{92}
\end{equation}
$$

We find the energy from the integral of the energy density over the volume between the inner and the outer radius, where the field is non-zero. The energy is:

<!-- Equation labels as ordinary links -->
<div id="_auto91"></div>

$$
\begin{equation}
W_m = \int \frac{1}{2}\mathbf{B} \cdot \mathbf{H} \mathrm{d} v = \frac{1}{2}\int \mu H^2 \mathrm{d} v 
\label{_auto91} \tag{93}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto92"></div>

$$
\begin{equation}  
    = \frac{1}{2}\int \mu \frac{I^2}{(2 \pi r)^2} r \mathrm{d} \phi \mathrm{d} r \mathrm{d} 
\label{_auto92} \tag{94}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto93"></div>

$$
\begin{equation}  
    = l 2 \pi \int_a^b \frac{1}{2}\mu \frac{I^2}{(2 \pi)^2} \frac{1}{r} 
\label{_auto93} \tag{95}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto94"></div>

$$
\begin{equation}  
    = \frac{1}{2}l \frac{I^2}{2 \pi}\mu \ln \frac{b}{a} 
\label{_auto94} \tag{96}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto95"></div>

$$
\begin{equation}  
    = \frac{1}{2} l \mu \frac{I^2}{2 \pi} \ln \frac{b}{a} 
\label{_auto95} \tag{97}
\end{equation}
$$

<!-- Equation labels as ordinary links -->
<div id="_auto96"></div>

$$
\begin{equation}  
    = \frac{1}{2} l \underbrace{\frac{\mu}{2 \pi} \ln \frac{b}{a}}_{L} I^2 \; .
\label{_auto96} \tag{98}
\end{equation}
$$

In principle, we can therefore find $L$ by measuring the energy needed to charge the circuit and calculating $L$ from $W_m = \frac{1}{2}LI^2$.

# Magnetic forces

We have a way to find the force from a magnetic field on a current-carrying element, where we found that the contribution from a current element $I \mathrm{d} \mathbf{l}$ is $\mathrm{d} \mathbf{F} = I \mathrm{d} \mathbf{l} \times \mathbf{B}$. However, this is not practical in all situations. For example, if we want to find the force between two permament magnets we would need to know the magnetic field from one of them and the (bound) current density in the other to calculate the force. Here, we will introduce a more practical method based on energy considerations.

## Magnetic force in a loss-free system

Fig. [fig:inductance-magneticforce-010](#fig:inductance-magneticforce-010) illustrates a loss-free system with a permanent magnet moving relative to an ideal conductor. The sytem is closed, so that the energy is conserved. Assume we displace the magnet in the $z$-direction, $z \rightarrow z + \mathrm{d} z$. The work performed by the magnetic force is then:

<!-- Equation labels as ordinary links -->
<div id="_auto97"></div>

$$
\begin{equation}
\mathrm{d} W = \mathbf{F} \cdot \mathbf{\Delta \mathbf{r}} = \mathbf{F}_m \cdot \mathrm{d} z \mathbf{\hat{z}} = F_{m,z} \mathrm{d} z = - d W_m \; ,
\label{_auto97} \tag{99}
\end{equation}
$$

where $W_m$ is the magnetic energy (and not a work). We interpret this to mean that the work done by the magnetic force must lead to a change in the magnetic energy. We can write this as:

<!-- Equation labels as ordinary links -->
<div id="_auto98"></div>

$$
\begin{equation}
F_{m,s} = - \frac{\partial W_m}{\partial z} \; .
\label{_auto98} \tag{100}
\end{equation}
$$

We can introduce a similar argument for motion in the $x$ and $y$ directions. In general, we therefore get:

<!-- Equation labels as ordinary links -->
<div id="_auto99"></div>

$$
\begin{equation}
\mathbf{F}_m = - \nabla W_m \; ,
\label{_auto99} \tag{101}
\end{equation}
$$

where we must remember that $W_m$ is an energy (and not work). This is a practical formula, because we may be able to find $W_m(x,y,z)$ for the system.

## Example: Carrying force of an electromagnet

Fig. [fig:inductance-horseshoe-010](#fig:inductance-horseshoe-010) illustrates a horseshoe-shaped iron electromagnet with a wire wrapped around its top. When a current $I$ is running in the wire, the magnet is able to lift a weight $W$ made of iron. What is the lifting force of the electromagnet?

<!-- dom:FIGURE:[fig-inductance/inductance-horseshoe-010.png, width=600 frac=0.45] A horseshoe shaped iron core electromagnet lifting an iron bar. <div id="fig:inductance-horseshoe-010"></div> -->
<!-- begin figure -->
<div id="fig:inductance-horseshoe-010"></div>

<p>A horseshoe shaped iron core electromagnet lifting an iron bar.</p>
<img src="fig-inductance/inductance-horseshoe-010.png" width=600>

<!-- end figure -->


**Solution:**

Our plan is to find an expression for the magnetic energy $W_m$ of the system as function of the distance $\mathrm{d} z$ between the horseshoe magnet and the iron bar below it. We will assume that the permeability of the iron magnet is large ($\mu \rightarrow \infty$) so that all the flux remains inside the magnetic circuit consisting of the horseshoe magnet, the iron bar and the air gaps of thickness $\mathrm{d} z$. The total energy will consist of three part for the horseshoe, $W_{m,h}$, the air gaps, $W_{m,a}$ and the iron bar $W_{m,b}$. However, the energies of the horseshow and the iron bar will remain the same, it is only the magnetic energy in the air gaps that will depend on $\mathrm{d} z$.

$$
\begin{equation}
W_m = W_{m,h} + W_{m,a} + W_{m,b}
\label{}
\end{equation}
$$

We find that the magnetic energy in the air gaps is the energy density in the air times the volume of the air gaps. Each air gap has a cross-sectional area $S$ and a length $\mathrm{d} z$. The energy is therefore:

<!-- Equation labels as ordinary links -->
<div id="_auto100"></div>

$$
\begin{equation}
W_{m,a} = - \underbrace{\frac{B^2}{2 \mu_0}}_{u} \, \underbrace{2 S \mathrm{d} z }_{dv} = - \frac{B^2 S \mathrm{d} z}{\mu_0} \; .
\label{_auto100} \tag{102}
\end{equation}
$$

The resulting force is then

<!-- Equation labels as ordinary links -->
<div id="_auto101"></div>

$$
\begin{equation}
\mathbf{F} = F_z \mathbf{\hat{z}} = - \frac{B^2 S}{\mu_0}\; .
\label{_auto101} \tag{103}
\end{equation}
$$

The force is in the negative $z$-direction, that is, it is an attractive force pulling the two parts together. For a magnet with magnetic field $B = 1 \text{T}$ and a surface area of $1 \text{cm}^2$, we find that the force is $F \simeq 80 \text{N}$, which means that it can support about $80 / g \simeq 8 \text{kg}$.

# Summary

The **self inductance** of a system with a flux $\Phi$ when a current $I$ is running through it is $L = \Phi/I$.

We find the **self inductance** by assuming a current $I$ is running in the circuit, finding the resulting magnetic field $\mathbf{B}$ and the flux $\Phi$ of the magnetici field through the circuit. The self inductance $L$ is then found from $L = \Phi/I$. 

The flux *in* a circuit $C_j$ *from* a current $I_i$ in circuit $C_i$ is $\Phi_{i,j} = \int_{S_1} \mathbf{B}_i \cdot \mathrm{d} \mathbf{S}$, where $\mathbf{B}_i$ is the magnetic field from current $I_i$. The **mutual inductance** is defined as:

$$
L_{i,j} = \frac{\Phi_{i,j}}{I_i} \; .
$$

The inductance $L_{i,i}$ is the self-inductance of circuit $i$.

The **mutual inductances are symmetric**: $L_{i,j} = L_{j,j}$.

The **magnetic energy** $W_m$ stored in an inductor with inductance $L$ and current $I$ is

$$
W_m = \frac{1}{2} L I^2 = \frac{1}{2} I \Phi \; .
$$

The **energy density** $u_m$ of a magnetic field is

$$
u_m = \frac{1}{2} \mathbf{B} \cdot \mathbf{H} \; .
$$
